```r
df_DHS_distinct <- read_csv("./interm/df_distinct_cells_CDC.csv")
sprintf("%i x %i dataframe", nrow(df_DHS_distinct), ncol(df_DHS_distinct))
head(df_DHS_distinct,2)
```

## HI bin 1

```r
df_HI1 <- fread(".././0A_input/quantile_WB_bins/bin1.csv")
sprintf("%i x %i dataframe", nrow(df_HI1), ncol(df_HI1))
head(df_HI1,2)
```

```r
sum(df_HI1==1)
```

```r
df1 <- data.frame(df_DHS_distinct) %>% 
            left_join(df_HI1,by=c("cell_x"="x","cell_y"="y"))
head(df1,2)
```

```r
names1 <- colnames(df1)
head(names1)
```

```r
ncol1 <- ncol(df1)
ncol1
```

```r
# translator of measured date to column index
date2index <- 1:ncol(df1)
names(date2index) <- colnames(df1)
```

```r
for (h in 1:30){
    df1[ncol1+h] <- sapply(1:nrow(df1), 
                      function(i) df1[i, date2index[as.character(df1$Measured_date_CDC[i]-h)]])
}
```

```r
newnames1 <- c()
for (h in 1:30){
    newname <- paste0("HI1DayMin",h)
    newnames1 <- append(newnames1,newname)
}
```

```r
colnames(df1) <- append(names1,newnames1)
head(df1,2)
```

```r
rm(df_HI1)
```

## HI bin 4

# Final selection

```r
df_distinct_final <- df3 %>%
                        select(-starts_with("4"))%>%
                        #Remove missing values
                        filter(!is.na(HI1DayMin1))%>%
                        #Sum across columns
                        mutate(HI1_bin_M1 = rowSums(select(.,HI1DayMin1:HI1DayMin30)),
                               HI2_bin_M1 = rowSums(select(.,HI2DayMin1:HI2DayMin30)),
                               HI3_bin_M1 = rowSums(select(.,HI3DayMin1:HI3DayMin30)),
                               #HI4_bin_M1 = rowSums(select(.,HI4DayMin1:HI4DayMin30)),
                               HI1_bin_D2 = rowSums(select(.,HI1DayMin1:HI1DayMin2)),
                               HI2_bin_D2 = rowSums(select(.,HI2DayMin1:HI2DayMin2)),
                               HI3_bin_D2 = rowSums(select(.,HI3DayMin1:HI3DayMin2)),
                               #HI4_bin_D2 = rowSums(select(.,HI4DayMin1:HI4DayMin2)),
                               HI1_bin_D10 = rowSums(select(.,HI1DayMin1:HI1DayMin10)),
                               HI2_bin_D10 = rowSums(select(.,HI2DayMin1:HI2DayMin10)),
                               HI3_bin_D10 = rowSums(select(.,HI3DayMin1:HI3DayMin10)),
                               #HI4_bin_D10 = rowSums(select(.,HI4DayMin1:HI4DayMin10)),
                               HI1_bin_S1 = rowSums(select(.,HI1DayMin1:HI1DayMin7)),
                               HI2_bin_S1 = rowSums(select(.,HI2DayMin1:HI2DayMin7)),
                               HI3_bin_S1 = rowSums(select(.,HI3DayMin1:HI3DayMin7)),
                               #HI4_bin_S1 = rowSums(select(.,HI4DayMin1:HI4DayMin7)),
                               HI1_bin_S2 = rowSums(select(.,HI1DayMin1:HI1DayMin14)),
                               HI2_bin_S2 = rowSums(select(.,HI2DayMin1:HI2DayMin14)),
                               HI3_bin_S2 = rowSums(select(.,HI3DayMin1:HI3DayMin14)),
                               #HI4_bin_S2 = rowSums(select(.,HI4DayMin1:HI4DayMin14)),
                               HI1_bin_S3 = rowSums(select(.,HI1DayMin1:HI1DayMin21)),
                               HI2_bin_S3 = rowSums(select(.,HI2DayMin1:HI2DayMin21)),
                               HI3_bin_S3 = rowSums(select(.,HI3DayMin1:HI3DayMin21)),
                               #HI4_bin_S3 = rowSums(select(.,HI4DayMin1:HI4DayMin21))
                              )%>%
                        select(-contains("Min"))%>%
                        mutate(HI0_bin_M1 = 720 - rowSums(select(.,HI1_bin_M1:HI3_bin_M1)),
                               HI0_bin_D2 = 48 - rowSums(select(.,HI1_bin_D2:HI3_bin_D2)),
                               HI0_bin_D10 = 240 - rowSums(select(.,HI1_bin_D10:HI3_bin_D10)),
                               HI0_bin_S1 = 168 - rowSums(select(.,HI1_bin_S1:HI3_bin_S1)),
                               HI0_bin_S2 = 336 - rowSums(select(.,HI1_bin_S2:HI3_bin_S2)),
                               HI0_bin_S3 = 504 - rowSums(select(.,HI1_bin_S3:HI3_bin_S3)))
sprintf("%i x %i dataframe", nrow(df_distinct_final), ncol(df_distinct_final))
head(df_distinct_final,2)
```
