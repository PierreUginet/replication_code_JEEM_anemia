```r
list.of.packages <- c("tidyverse","data.table","stargazer","patchwork")
new.packages <- list.of.packages[!(list.of.packages %in% installed.packages()[,"Package"])]
if(length(new.packages)) install.packages(new.packages, repos = "http://cran.us.r-project.org")

invisible(lapply(list.of.packages, library, character.only = TRUE))

options(repr.matrix.max.rows=500, repr.matrix.max.cols=1200, message=FALSE, warning=TRUE)  
```

```r
df_Ti30 <- read_csv("./output/df_T_cells_CDC_30_20_10.csv")
sprintf("%i x %i dataframe", nrow(df_Ti30), ncol(df_Ti30))
head(df_Ti30,2)
```

```r
as.Date(43828, origin = "1900-01-01")
```

```r
T_bins_labels <- c("<=17°C","(17,21]","(21,23]","(23,25]","(25,27]","(27,29]","(29,31]","(31,33]",">33°C")
```

```r
df_Ti30_ok <- df_Ti30%>%
                filter(!is.na(T_i30) & Measured_date_CDC <= 43828
                      )%>%
                mutate(T_i30_bin=factor(T_i30_bin,levels = T_bins_labels),
                       T_i20_bin=factor(T_i20_bin,levels = T_bins_labels),
                       T_i10_bin=factor(T_i10_bin,levels = T_bins_labels))
sprintf("%i x %i dataframe", nrow(df_Ti30_ok), ncol(df_Ti30_ok))
head(df_Ti30_ok,2)
```

## Calculate deciles

```r
deciles <- quantile(df_Ti30_ok$T_i30, probs = seq(0, 1, by = 0.1))
deciles
```

## Plot histogram per 1°C

```r
theme_perso <- function(...) {
                    theme_bw()+
                    theme(legend.position="none",
                          plot.title = element_text(size=30),
                          axis.title.x = element_text(size = 22, face ="bold"),
                          axis.text.x =  element_text(size = 20, angle=45, hjust=1),
                          axis.title.y = element_text(size = 22, face ="bold"),
                          axis.text.y =  element_text(size = 20)
                         )
}
```

```r
options(repr.plot.width=7, repr.plot.height=7, scipen=9999)

plot <- ggplot(df_Ti30_ok, aes(x = T_i30)) +
              geom_histogram(fill = "lightblue", 
                             color = "black",
                             breaks = seq(-15, 40, by = 1)) +
              geom_vline(xintercept = deciles, color = "red", linetype = "dashed", linewidth = 0.5) +
              labs(title = "",
                     x = "\n Air temperature (°C)",
                     y = "Number of grid cells \n") +
              xlim(-15, 40) +
              theme_minimal() + 
              theme_perso()
plot
```

```r
ggsave("./output/img/histo_deciles_T.png", width=7, height=7, plot)
```

```r
data.frame(df_Ti30_ok) %>% 
        select(T_i30,T_i20,T_i10)%>%
        stargazer(nobs=TRUE, digits = 2,
                  summary.stat = c("n","mean", "sd","min","p25","median","p75","max"),
                  title = c("Summary statistics"),
                  type="text"
                     )
```

## Plot bins histograms

```r
df_Ti30_boxplot <- df_Ti30_ok %>%
                        group_by(T_i30_bin)%>%
                        summarize(value=n())
head(df_Ti30_boxplot,2)
```

```r
df_Ti20_boxplot <- df_Ti30_ok %>%
                        group_by(T_i20_bin)%>%
                        summarize(value=n())
head(df_Ti20_boxplot,2)
```

```r
df_Ti10_boxplot <- df_Ti30_ok %>%
                        group_by(T_i10_bin)%>%
                        summarize(value=n())
head(df_Ti10_boxplot,2)
```

```r
T_plot_mean_30 <- ggplot(df_Ti30_boxplot, aes(x=T_i30_bin, y=value)) +
                geom_bar(stat="identity") +
                theme_perso() +
                ggtitle("30 days") +
                labs(y="Number of grid cells \n")+
                scale_x_discrete(name="\n Air temperature bin (°C)")
```

```r
T_plot_mean_20 <- ggplot(df_Ti20_boxplot, aes(x=T_i20_bin, y=value)) +
                geom_bar(stat="identity") +
                theme_perso() +
                ggtitle("20 days") +
                labs(y="Number of grid cells \n")+
                scale_x_discrete(name="\n Air temperature bin (°C)"
                                )
```

```r
T_plot_mean_10 <- ggplot(df_Ti10_boxplot, aes(x=T_i10_bin, y=value)) +
                geom_bar(stat="identity") +
                theme_perso() +
                ggtitle("10 days") +
                labs(y="Number of grid cells \n")+
                scale_x_discrete(name="\n Air temperature bin (°C)"
                                )
```

```r
options(repr.plot.width=30, repr.plot.height=10, scipen=9999)

plot_Ti <- T_plot_mean_30 + plot_spacer() + 
           T_plot_mean_20 + plot_spacer() + 
           T_plot_mean_10 + 
           plot_layout(widths=c(10,1,10,1,10), nrow=1)
plot_Ti
```

```r
ggsave("./output/img/histo_Ti30_20_10.png", width=30, height=10, plot_Ti)
```
