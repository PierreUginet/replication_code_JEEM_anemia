## Dataframe

```r
df_N <- read_csv("./output/df_WB_cells_CDC_Nhours.csv")
sprintf("%i x %i dataframe", nrow(df_N), ncol(df_N))
head(df_N,2)
```

## Mean temperature

```r
df_reg_boxplot_T_30d <- df_N %>%
                         select(WB_Bin_inf14_Nhours_on30,WB_Bin_14_16_Nhours_on30,WB_Bin_16_18_Nhours_on30,
                                WB_Bin_18_20_Nhours_on30,WB_Bin_20_22_Nhours_on30,WB_Bin_22_24_Nhours_on30,
                                WB_Bin_24_26_Nhours_on30,WB_Bin_26_28_Nhours_on30,WB_Bin_sup28_Nhours_on30)%>%
                         pivot_longer(cols = WB_Bin_inf14_Nhours_on30:WB_Bin_sup28_Nhours_on30, 
                                      names_to = "WB_bin",
                                      values_to = "value"
                                        )%>%
                        mutate(WB_bin=factor(WB_bin,levels = c("WB_Bin_inf14_Nhours_on30","WB_Bin_14_16_Nhours_on30",
                                                               "WB_Bin_16_18_Nhours_on30","WB_Bin_18_20_Nhours_on30",
                                                               "WB_Bin_20_22_Nhours_on30","WB_Bin_22_24_Nhours_on30",
                                                               "WB_Bin_24_26_Nhours_on30","WB_Bin_26_28_Nhours_on30", 
                                                               "WB_Bin_sup28_Nhours_on30")))
sprintf("%i rows x %i columns", nrow(df_reg_boxplot_T_30d), ncol(df_reg_boxplot_T_30d))
head(df_reg_boxplot_T_30d,2)
```

```r
df_reg_boxplot_T_20d <- df_N %>%
                         select(WB_Bin_inf14_Nhours_on20,WB_Bin_14_16_Nhours_on20,WB_Bin_16_18_Nhours_on20,
                                WB_Bin_18_20_Nhours_on20,WB_Bin_20_22_Nhours_on20,WB_Bin_22_24_Nhours_on20,
                                WB_Bin_24_26_Nhours_on20,WB_Bin_26_28_Nhours_on20,WB_Bin_sup28_Nhours_on20)%>%
                         pivot_longer(cols = WB_Bin_inf14_Nhours_on20:WB_Bin_sup28_Nhours_on20, 
                                      names_to = "WB_bin",
                                      values_to = "value"
                                        )%>%
                        mutate(WB_bin=factor(WB_bin,levels = c("WB_Bin_inf14_Nhours_on20","WB_Bin_14_16_Nhours_on20",
                                                               "WB_Bin_16_18_Nhours_on20","WB_Bin_18_20_Nhours_on20",
                                                               "WB_Bin_20_22_Nhours_on20","WB_Bin_22_24_Nhours_on20",
                                                               "WB_Bin_24_26_Nhours_on20","WB_Bin_26_28_Nhours_on20", 
                                                               "WB_Bin_sup28_Nhours_on20")))
sprintf("%i rows x %i columns", nrow(df_reg_boxplot_T_20d), ncol(df_reg_boxplot_T_20d))
head(df_reg_boxplot_T_20d,2)
```

```r
df_reg_boxplot_T_10d <- df_N %>%
                         select(WB_Bin_inf14_Nhours_on10,WB_Bin_14_16_Nhours_on10,WB_Bin_16_18_Nhours_on10,
                                WB_Bin_18_20_Nhours_on10,WB_Bin_20_22_Nhours_on10,WB_Bin_22_24_Nhours_on10,
                                WB_Bin_24_26_Nhours_on10,WB_Bin_26_28_Nhours_on10,WB_Bin_sup28_Nhours_on10)%>%
                         pivot_longer(cols = WB_Bin_inf14_Nhours_on10:WB_Bin_sup28_Nhours_on10, 
                                      names_to = "WB_bin",
                                      values_to = "value"
                                        )%>%
                        mutate(WB_bin=factor(WB_bin,levels = c("WB_Bin_inf14_Nhours_on10","WB_Bin_14_16_Nhours_on10",
                                                               "WB_Bin_16_18_Nhours_on10","WB_Bin_18_20_Nhours_on10",
                                                               "WB_Bin_20_22_Nhours_on10","WB_Bin_22_24_Nhours_on10",
                                                               "WB_Bin_24_26_Nhours_on10","WB_Bin_26_28_Nhours_on10", 
                                                               "WB_Bin_sup28_Nhours_on10")))
sprintf("%i rows x %i columns", nrow(df_reg_boxplot_T_10d), ncol(df_reg_boxplot_T_10d))
head(df_reg_boxplot_T_10d,2)
```

## Plot

```r
theme_perso <- function(...) {
                    theme_bw()+
                    theme(legend.position="none",
                          plot.title = element_text(size=30),
                          axis.title.x = element_text(size = 22, face ="bold"),
                          axis.text.x =  element_text(size = 20, angle=45, hjust=1),
                          axis.title.y = element_text(size = 22, face ="bold"),
                          axis.text.y =  element_text(size = 20)
                         )
}
```

```r
T_plot_30 <- ggplot(df_reg_boxplot_T_30d, aes(x=WB_bin, y=value)) +
                geom_boxplot() +
                theme_perso() +
                ggtitle("30 days") +
                labs(y="Number of hours \n")+
                scale_x_discrete(name="\n WBGT bin (°C)",
                                 labels=c("<=14°C","(14,16]","(16,18]","(18,20]","(20,22]","(22,24]","(24,26]","(26,28]",">28°C")
                                )
```

```r
T_plot_20 <- ggplot(df_reg_boxplot_T_20d, aes(x=WB_bin, y=value)) +
                geom_boxplot() +
                theme_perso() +
                ggtitle("20 days") +
                labs(y="Number of hours \n")+
                scale_x_discrete(name="\n WBGT bin (°C)",
                                 labels=c("<=14°C","(14,16]","(16,18]","(18,20]","(20,22]","(22,24]","(24,26]","(26,28]",">28°C")
                                )
```

```r
T_plot_10 <- ggplot(df_reg_boxplot_T_10d, aes(x=WB_bin, y=value)) +
                geom_boxplot() +
                theme_perso() +
                ggtitle("10 days") +
                labs(y="Number of hours \n")+
                scale_x_discrete(name="\n WBGT bin (°C)",
                                 labels=c("<=14°C","(14,16]","(16,18]","(18,20]","(20,22]","(22,24]","(24,26]","(26,28]",">28°C")
                                )
```

```r
options(repr.plot.width=30, repr.plot.height=10, scipen=9999)

plot_Nhours <- T_plot_30 + plot_spacer() + T_plot_20 + plot_spacer() + T_plot_10 +
              plot_layout(widths=c(10,1,10,1,10), nrow=1)
plot_Nhours
```

```r
ggsave("./output/img/boxplot_Nhours_WB30_20_10.png", width=30, height=10, plot_Nhours)
```
