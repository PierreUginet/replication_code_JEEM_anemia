## Dataframe

```r
df_N <- read_csv("./output/df_T_cells_CDC_Nhours.csv")
sprintf("%i x %i dataframe", nrow(df_N), ncol(df_N))
head(df_N,2)
```

## Mean temperature

```r
df_reg_boxplot_T_30d <- df_N %>%
                         select(T_Bin_inf17_Nhours_on30,T_Bin_17_21_Nhours_on30,T_Bin_21_23_Nhours_on30,
                                T_Bin_23_25_Nhours_on30,T_Bin_25_27_Nhours_on30,T_Bin_27_29_Nhours_on30,
                                T_Bin_29_31_Nhours_on30,T_Bin_31_33_Nhours_on30,T_Bin_sup33_Nhours_on30)%>%
                         pivot_longer(cols = T_Bin_inf17_Nhours_on30:T_Bin_sup33_Nhours_on30, 
                                      names_to = "T_bin",
                                      values_to = "value"
                                        )%>%
                        mutate(T_bin=factor(T_bin,levels = c("T_Bin_inf17_Nhours_on30","T_Bin_17_21_Nhours_on30",
                                                             "T_Bin_21_23_Nhours_on30","T_Bin_23_25_Nhours_on30",
                                                             "T_Bin_25_27_Nhours_on30","T_Bin_27_29_Nhours_on30", 
                                                             "T_Bin_29_31_Nhours_on30","T_Bin_31_33_Nhours_on30", 
                                                             "T_Bin_sup33_Nhours_on30")))
sprintf("%i rows x %i columns", nrow(df_reg_boxplot_T_30d), ncol(df_reg_boxplot_T_30d))
head(df_reg_boxplot_T_30d,2)
```

```r
df_reg_boxplot_T_20d <- df_N %>%
                         select(T_Bin_inf17_Nhours_on20,T_Bin_17_21_Nhours_on20,T_Bin_21_23_Nhours_on20,
                                T_Bin_23_25_Nhours_on20,T_Bin_25_27_Nhours_on20,T_Bin_27_29_Nhours_on20,
                                T_Bin_29_31_Nhours_on20,T_Bin_31_33_Nhours_on20,T_Bin_sup33_Nhours_on20)%>%
                         pivot_longer(cols = T_Bin_inf17_Nhours_on20:T_Bin_sup33_Nhours_on20, 
                                      names_to = "T_bin",
                                      values_to = "value"
                                        )%>%
                        mutate(T_bin=factor(T_bin,levels = c("T_Bin_inf17_Nhours_on20","T_Bin_17_21_Nhours_on20",
                                                             "T_Bin_21_23_Nhours_on20","T_Bin_23_25_Nhours_on20",
                                                             "T_Bin_25_27_Nhours_on20","T_Bin_27_29_Nhours_on20", 
                                                             "T_Bin_29_31_Nhours_on20","T_Bin_31_33_Nhours_on20", 
                                                             "T_Bin_sup33_Nhours_on20")))
sprintf("%i rows x %i columns", nrow(df_reg_boxplot_T_20d), ncol(df_reg_boxplot_T_20d))
head(df_reg_boxplot_T_20d,2)
```

```r
df_reg_boxplot_T_10d <- df_N %>%
                         select(T_Bin_inf17_Nhours_on10,T_Bin_17_21_Nhours_on10,T_Bin_21_23_Nhours_on10,
                                T_Bin_23_25_Nhours_on10,T_Bin_25_27_Nhours_on10,T_Bin_27_29_Nhours_on10,
                                T_Bin_29_31_Nhours_on10,T_Bin_31_33_Nhours_on10,T_Bin_sup33_Nhours_on10)%>%
                         pivot_longer(cols = T_Bin_inf17_Nhours_on10:T_Bin_sup33_Nhours_on10, 
                                      names_to = "T_bin",
                                      values_to = "value"
                                        )%>%
                        mutate(T_bin=factor(T_bin,levels = c("T_Bin_inf17_Nhours_on10","T_Bin_17_21_Nhours_on10",
                                                             "T_Bin_21_23_Nhours_on10","T_Bin_23_25_Nhours_on10",
                                                             "T_Bin_25_27_Nhours_on10","T_Bin_27_29_Nhours_on10", 
                                                             "T_Bin_29_31_Nhours_on10","T_Bin_31_33_Nhours_on10", 
                                                             "T_Bin_sup33_Nhours_on10")))
sprintf("%i rows x %i columns", nrow(df_reg_boxplot_T_10d), ncol(df_reg_boxplot_T_10d))
head(df_reg_boxplot_T_10d,2)
```

## Plot

```r
theme_perso <- function(...) {
                    theme_bw()+
                    theme(legend.position="none",
                          plot.title = element_text(size=30),
                          axis.title.x = element_text(size = 22, face ="bold"),
                          axis.text.x =  element_text(size = 20, angle=45, hjust=1),
                          axis.title.y = element_text(size = 22, face ="bold"),
                          axis.text.y =  element_text(size = 20)
                         )
}
```

```r
labels_temp <- c("<=17°C","(17,21]","(21,23]","(23,25]","(25,27]","(27,29]","(29,31]","(31,33]",">33°C")
```

```r
T_plot_30 <- ggplot(df_reg_boxplot_T_30d, aes(x=T_bin, y=value)) +
                geom_boxplot() +
                theme_perso() +
                ggtitle("30 days") +
                labs(y="Number of hours \n")+
                scale_x_discrete(name="\n Air temperature bin (°C)",
                                 labels=labels_temp
                                )
```

```r
T_plot_20 <- ggplot(df_reg_boxplot_T_20d, aes(x=T_bin, y=value)) +
                geom_boxplot() +
                theme_perso() +
                ggtitle("20 days") +
                labs(y="Number of hours \n")+
                scale_x_discrete(name="\n Air temperature bin (°C)",
                                 labels=labels_temp
                                )
```

```r
T_plot_10 <- ggplot(df_reg_boxplot_T_10d, aes(x=T_bin, y=value)) +
                geom_boxplot() +
                theme_perso() +
                ggtitle("10 days") +
                labs(y="Number of hours \n")+
                scale_x_discrete(name="\n Air temperature bin (°C)",
                                 labels=labels_temp
                                )
```

```r
options(repr.plot.width=30, repr.plot.height=10, scipen=9999)

plot_Nhours <- T_plot_30 + plot_spacer() + T_plot_20 + plot_spacer() + T_plot_10 +
              plot_layout(widths=c(10,1,10,1,10), nrow=1)
plot_Nhours
```

```r
ggsave("./output/img/boxplot_Nhours_T30_20_10.png", width=30, height=10, plot_Nhours)
```
