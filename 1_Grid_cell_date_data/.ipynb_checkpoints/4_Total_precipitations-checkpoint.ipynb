# Distinct cells and dates

```r
df_DHS_distinct <- read_csv("./interm/df_distinct_cells_CDC_025.csv")
sprintf("%i x %i dataframe", nrow(df_DHS_distinct), ncol(df_DHS_distinct))
head(df_DHS_distinct,2)
```

# Daily precipitations

```r
df_precip <- fread(".././0_input/Total_Precip/daily_prec.csv")
sprintf("%i x %i dataframe", nrow(df_precip), ncol(df_precip))
head(df_precip,2)
```

## Transform column names

```r
col_names <- colnames(df_precip)
head(col_names)
```

```r
class(col_names)
```

```r
col_names_s <- col_names[- c(1,2)] #remove first two elements
head(col_names_s)
```

```r
new_names <- c("cell_x","cell_y",as.numeric(as.Date(substr(col_names_s,1,10)) - as.Date("1900-01-01")))
head(new_names)
```

```r
colnames(df_precip) <- new_names
head(df_precip,2)
```

## Left_join precip to cells and date

```r
df_DHS_precip <- df_DHS_distinct %>%
                        left_join(df_precip,by=c("cell_x","cell_y"))
sprintf("%i x %i dataframe", nrow(df_DHS_precip), ncol(df_DHS_precip))
head(df_DHS_precip,2)
```

```r
date_cols <- setdiff(colnames(df_DHS_precip), c("Measured_date_CDC","cell_x","cell_y"))
head(date_cols)
```

```r
mat <- as.matrix(df_DHS_precip [, date_cols])
head(mat)
```

```r
df_DHS_precip$P_tot_30 <- sapply(seq_len(nrow(df_DHS_precip)), function(i) {
  cols_needed <- as.character(
    (df_DHS_precip$Measured_date_CDC[i] - 29):(df_DHS_precip$Measured_date_CDC[i])
  )
  
  valid_cols <- intersect(cols_needed, date_cols)
  
  sum(mat[i, match(valid_cols, date_cols)], na.rm = TRUE)
})
```

```r
df_DHS_precip$P_tot_20 <- sapply(seq_len(nrow(df_DHS_precip)), function(i) {
  cols_needed <- as.character(
    (df_DHS_precip$Measured_date_CDC[i] - 19):(df_DHS_precip$Measured_date_CDC[i])
  )
  
  valid_cols <- intersect(cols_needed, date_cols)
  
  sum(mat[i, match(valid_cols, date_cols)], na.rm = TRUE)
})
```

```r
df_DHS_precip$P_tot_10 <- sapply(seq_len(nrow(df_DHS_precip)), function(i) {
  cols_needed <- as.character(
    (df_DHS_precip$Measured_date_CDC[i] - 9):(df_DHS_precip$Measured_date_CDC[i])
  )
  
  valid_cols <- intersect(cols_needed, date_cols)
  
  sum(mat[i, match(valid_cols, date_cols)], na.rm = TRUE)
})
```

```r
df_sum_precip <- df_DHS_precip %>%
                        select(-starts_with("4"))%>%
                        filter(!is.na(P_tot_30))
sprintf("%i x %i dataframe", nrow(df_sum_precip), ncol(df_sum_precip))
head(df_sum_precip,2)
```

## SAVE

```r
fwrite(df_sum_precip,"./output/df_precip_cells_CDC_30_20_10.csv")
```
