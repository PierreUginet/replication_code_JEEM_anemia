## Plot

```r
theme_perso <- function(...) {
                    theme_bw()+
                    theme(legend.position="none",
                          plot.title = element_text(size=30),
                          axis.title.x = element_text(size = 22, face ="bold"),
                          axis.text.x =  element_text(size = 20, angle=45, hjust=1),
                          axis.title.y = element_text(size = 22, face ="bold"),
                          axis.text.y =  element_text(size = 20)
                         )
}
```

```r
labels_temp_min <- c("≤ 11°C","(11,13]","(13,15]","(15,17]","(15,19]","(17,21]","(21,23]","(23,25]",">25°C")
labels_temp_mean <- c("≤ 17°C","(17,21]","(21,23]","(23,25]","(25,27]","(27,29]","(29,31]","(31,33]",">33°C")
labels_temp_max <- c("≤ 21°C","(21,23]","(23,25]","(25,27]","(27,29]","(29,31]","(31,33]","(33,35]",">37°C")
```

```r
T_plot_min_30 <- ggplot(df_reg_boxplot_T_min_30d, aes(x=T_bin, y=value)) +
                geom_boxplot() +
                theme_perso() +
                ggtitle("30 days") +
                labs(y="Number of days \n")+
                scale_x_discrete(name="\n Min Air temperature bin (°C)",
                                 labels=labels_temp_min
                                )
```

```r
T_plot_mean_30 <- ggplot(df_reg_boxplot_T_mean_30d, aes(x=T_bin, y=value)) +
                geom_boxplot() +
                theme_perso() +
                ggtitle("30 days") +
                labs(y="Number of days \n")+
                scale_x_discrete(name="\n Mean Air temperature bin (°C)",
                                 labels=labels_temp_mean
                                )
```

```r
T_plot_max_30 <- ggplot(df_reg_boxplot_T_max_30d, aes(x=T_bin, y=value)) +
                geom_boxplot() +
                theme_perso() +
                ggtitle("30 days") +
                labs(y="Number of days \n")+
                scale_x_discrete(name="\n Max Air temperature bin (°C)",
                                 labels=labels_temp_max
                                )
```
