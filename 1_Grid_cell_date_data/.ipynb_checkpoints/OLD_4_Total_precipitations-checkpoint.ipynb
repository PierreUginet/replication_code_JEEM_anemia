# Distinct cells and dates

```r
df_DHS_distinct <- read_csv("./interm/df_distinct_cells_CDC.csv")
sprintf("%i x %i dataframe", nrow(df_DHS_distinct), ncol(df_DHS_distinct))
head(df_DHS_distinct,2)
```

# Daily precipitations

```r
df_precip <- fread(".././0A_input/Total_Precip/TP_daily_levels.csv")
sprintf("%i x %i dataframe", nrow(df_precip), ncol(df_precip))
head(df_precip,2)
```

## Transform column names

```r
col_names <- colnames(df_precip)
head(col_names)
```

```r
class(col_names)
```

```r
col_names_s <- col_names[- c(1,2)] #remove first two elements
head(col_names_s)
```

```r
new_names <- c("cell_x","cell_y",substr(col_names_s,6,10))
head(new_names)
```

```r
colnames(df_precip) <- new_names
head(df_precip,2)
```

## Left_join precip to cells and date

```r
df_DHS_precip <- df_DHS_distinct %>%
                        left_join(df_precip,by=c("cell_x","cell_y"))%>%
                        mutate(P_tot_30 = rowSums(select(.,paste(Measured_date_CDC-29):paste(Measured_date_CDC))),
                               P_tot_20 = rowSums(select(.,paste(Measured_date_CDC-19):paste(Measured_date_CDC))),
                               P_tot_10 = rowSums(select(.,paste(Measured_date_CDC-9):paste(Measured_date_CDC))))%>%
                        select(-starts_with("4"))%>%
                        filter(!is.na(P_tot_30))
sprintf("%i x %i dataframe", nrow(df_DHS_precip), ncol(df_DHS_precip))
head(df_DHS_precip,2)
```

## SAVE

```r
fwrite(df_DHS_precip,"./output/df_precip_cells_CDC_30_20_10.csv")
```
