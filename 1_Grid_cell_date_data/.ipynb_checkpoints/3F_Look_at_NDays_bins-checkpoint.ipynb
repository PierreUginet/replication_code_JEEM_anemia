## Plot

```r
theme_perso <- function(...) {
                    theme_bw()+
                    theme(legend.position="none",
                          plot.title = element_text(size=30),
                          axis.title.x = element_text(size = 22, face ="bold"),
                          axis.text.x =  element_text(size = 20, angle=45, hjust=1),
                          axis.title.y = element_text(size = 22, face ="bold"),
                          axis.text.y =  element_text(size = 20)
                         )
}
```

```r
labels_temp_min <- c("≤ 10°C","(10,12]","(12,14]","(14,16]","(16,18]","(18,20]","(20,22]","(22,24]",">24°C")
labels_temp_mean <- c("≤ 14°C","(14,16]","(16,18]","(18,20]","(20,22]","(22,24]","(24,26]","(26,28]",">28°C")
labels_temp_max <- c("≤ 18°C","(18,20]","(20,22]","(22,24]","(24,26]","(26,28]","(28,30]","(30,32]",">32°C")
```

```r
T_plot_min_30 <- ggplot(df_reg_boxplot_T_min_30d, aes(x=WB_bin, y=value)) +
                geom_boxplot() +
                theme_perso() +
                ggtitle("N 30 days - Min WBGT boxplot") +
                labs(y="Days in WB bin \n")+
                scale_x_discrete(name="\n Min WBGT bin",
                                 labels=labels_temp_min
                                )
```

```r
T_plot_mean_30 <- ggplot(df_reg_boxplot_T_mean_30d, aes(x=WB_bin, y=value)) +
                geom_boxplot() +
                theme_perso() +
                ggtitle("N 30 days - Mean WBGT boxplot") +
                labs(y="Days in WB bin \n")+
                scale_x_discrete(name="\n Mean WBGT bin",
                                 labels=labels_temp_mean
                                )
```

```r
T_plot_max_30 <- ggplot(df_reg_boxplot_T_max_30d, aes(x=WB_bin, y=value)) +
                geom_boxplot() +
                theme_perso() +
                ggtitle("N 30 days - Max WBGT boxplot") +
                labs(y="Days in WB bin \n")+
                scale_x_discrete(name="\n Max WBGT bin",
                                 labels=labels_temp_max
                                )
```
