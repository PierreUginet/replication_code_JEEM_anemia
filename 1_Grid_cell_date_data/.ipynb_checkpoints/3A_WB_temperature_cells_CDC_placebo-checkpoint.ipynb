# Distinct cells and dates

```r
df_DHS_distinct <- read_csv("./interm/df_distinct_cells_CDC.csv")
sprintf("%i x %i dataframe", nrow(df_DHS_distinct), ncol(df_DHS_distinct))
head(df_DHS_distinct,2)
```

# Temperature bins

## Function positive temperature

```r
temp_bins_pos_N_days <- function(N_deg_bins){
    df_cells_date <- data.frame(df_DHS_distinct)
    for (i in 0:N_deg_bins){
        print(i)
        df_Ti <- fread(paste0(".././0A_input/one_degree_WB_bins/bin",i,".csv"))
        df_i <- df_cells_date %>% 
                    left_join(df_Ti,by=c("cell_x"="x","cell_y"="y"))
        names_i <- colnames(df_i)
        ncol_i <- ncol(df_i)
        # translator of measured date to column index
        date2index <- 1:ncol(df_i)
        names(date2index) <- colnames(df_i)
        #Identify 30 days prior anemia measurement
        for (h in 1:30){
            df_i[ncol_i+h] <- sapply(1:nrow(df_i), 
                              function(n) df_i[n, date2index[as.character(df_i$Measured_date_CDC[n]-h)]])
        }
        #Change colnames
        newnames_i <- c()
        for (h in 1:30){
            newname <- paste0("T",i,"DayMin",h)
            newnames_i <- append(newnames_i,newname)
        }
        colnames(df_i) <- append(names_i,newnames_i)
        #Keep memory
        rm(df_Ti)
        #Update dataframe
        df_cells_date <- df_i  %>%
                            select(-starts_with("4"))
    }
    return(df_cells_date)
}
```

```r
df_deg_bins_pos <- temp_bins_pos_N_days(34)
sprintf("%i x %i dataframe", nrow(df_deg_bins_pos), ncol(df_deg_bins_pos))
head(df_deg_bins_pos,2)
```

```r
write_csv(df_deg_bins_pos,"./interm/df_WB_cells_CDC_positive.csv")
```

## Function negative temperature

```r
temp_bins_neg_N_days <- function(N_deg_bins){
    df_cells_date <- data.frame(df_DHS_distinct)
    for (i in 1:N_deg_bins){
        print(i)
        df_Ti <- fread(paste0(".././0A_input/one_degree_WB_bins/bin",-i,".csv"))
        df_i <- df_cells_date %>% 
                    left_join(df_Ti,by=c("cell_x"="x","cell_y"="y"))
        names_i <- colnames(df_i)
        ncol_i <- ncol(df_i)
        # translator of measured date to column index
        date2index <- 1:ncol(df_i)
        names(date2index) <- colnames(df_i)
        #Identify 30 days prior anemia measurement
        for (h in 1:30){
            df_i[ncol_i+h] <- sapply(1:nrow(df_i), 
                              function(n) df_i[n, date2index[as.character(df_i$Measured_date_CDC[n]-h)]])
        }
        #Change colnames
        newnames_i <- c()
        for (h in 1:30){
            newname <- paste0("T",-i,"DayMin",h)
            newnames_i <- append(newnames_i,newname)
        }
        colnames(df_i) <- append(names_i,newnames_i)
        #Keep memory
        rm(df_Ti)
        #Update dataframe
        df_cells_date <- df_i  %>%
                            select(-starts_with("4"))
    }
    return(df_cells_date)
}
```

```r
df_deg_bins_neg <- temp_bins_neg_N_days(30)
sprintf("%i x %i dataframe", nrow(df_deg_bins_neg), ncol(df_deg_bins_neg))
head(df_deg_bins_neg,2)
```
