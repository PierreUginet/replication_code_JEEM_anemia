# Datasets

```r
df_children <- fread("./2_Regression_datasets/output/df_reg_children.csv")
sprintf("%i rows x %i columns", nrow(df_children), ncol(df_children))
head(df_children,2)
```

```r
df_men <- fread("./2_Regression_datasets/output/df_reg_men.csv")
sprintf("%i rows x %i columns", nrow(df_men), ncol(df_men))
head(df_men,2)
```

```r
df_women <- fread("./2_Regression_datasets/output/df_reg_women.csv")
sprintf("%i rows x %i columns", nrow(df_women), ncol(df_women))
head(df_women,2)
```

# Plots

```r
options(repr.plot.width=12, repr.plot.height=8, scipen=10000)

plot_children <- ggplot(df_children, aes(x=Child_hemo_level_alti)) +
                                geom_density(#fill="#69b3a2", color="#e9ecef", alpha=0.8
                                            )+
                                labs(title="Children",y="",
                                     x="Hemoglobin level adjusted for altitude (g/L)")+
                                coord_cartesian(x=c(0,260),y=c(0,0.035))+
                                geom_vline(xintercept=110, linetype="dashed", color = "red")+
                                annotate("text", x = 60, y = 0.022, 
                                         label = "Anemia \n hemo <110g/L",
                                         angle = 0, color = "red", size = 7)+
                                ggtitle("Children")+
                                theme(plot.title = element_text(size=30, face = "bold"),
                                      axis.title.x = element_text(size = 22),
                                      axis.text.x =  element_text(size = 20),
                                      axis.text.y =  element_text(size = 20))
plot_children
```

```r
options(repr.plot.width=12, repr.plot.height=8, scipen=10000)

plot_men <- ggplot(df_men, aes(x=Blood_hemo_level_alti)) +
                                geom_density(#fill="#69b3a2", color="#e9ecef", alpha=0.8
                                            )+
                                labs(title="Men",y="",
                                     x="Hemoglobin level adjusted for altitude (g/L)")+
                                coord_cartesian(x=c(0,260),y=c(0,0.035))+
                                geom_vline(xintercept=130, linetype="dashed", color = "red")+
                                annotate("text", x = 60, y = 0.022, 
                                         label = "Anemia \n hemo <130g/L",
                                         angle = 0, color = "red", size=7)+
                                theme(plot.title = element_text(size=30, face = "bold"),
                                      axis.title.x = element_text(size = 22),
                                      axis.text.x =  element_text(size = 20),
                                      axis.text.y =  element_text(size = 20))
plot_men
```

```r
options(repr.plot.width=12, repr.plot.height=8, scipen=10000)

plot_women <- ggplot(df_women, aes(x=Blood_hemo_level_alti, color=as.factor(Currently_pregnant))) +
                                geom_density()+
                                scale_color_manual(values=c("black","brown"))+
                                labs(title="Women",y="",
                                     x="Hemoglobin level adjusted for altitude (g/L)")+
                                coord_cartesian(x=c(0,260),y=c(0,0.035))+
                                geom_vline(xintercept=110, linetype="dashed", color = "brown")+
                                annotate("text", x = 60, y = 0.015, 
                                         label = "Anemia \n pregnant woman \n hemo <110g/L",
                                         angle = 0, color = "brown", size=7)+
                                geom_vline(xintercept=120, linetype="dashed", color = "red")+
                                annotate("text", x = 60, y = 0.025, 
                                         label = "Anemia \n non-pregnant woman \n hemo <120g/L",
                                         angle = 0, color = "red", size=7)+
                                theme(plot.title = element_text(size=30, face = "bold"),
                                      axis.title.x = element_text(size = 22),
                                      axis.text.x =  element_text(size = 20),
                                      axis.text.y =  element_text(size = 20),
                                      legend.position="none")
plot_women
```

```r
options(repr.plot.width=12, repr.plot.height=20, scipen=10000)

plot_all <- ggarrange(plot_children,plot_women,plot_men,
                      nrow=3)
plot_all
```

```r
ggsave("./3_output/img/hemoglo_density.png", width=12, height=20, plot_all)
```
