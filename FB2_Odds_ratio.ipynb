# Number of people suffering from anemia

### Children

```r
reg_children <- glmer(Children_anemia ~ T_i30 +
                                        P_tot_30 +
                                        Child_female + Child_birth_order + Child_alive_age_month + 
                                        Mother_prim_educ + Mother_second_educ + Mother_higher_educ +
                                        Mother_muslim + Mother_not_hindu_nor_muslim  + 
                                        Mother_ethni_SC + Mother_ethni_ST + Mother_ethni_OBC +
                                        Wealth_lowest + Wealth_second + Wealth_fourth + Wealth_highest + 
                                        Urban +
                                        Measured_month +
                                        Measured_year +
                                        (1|District_ID),
                  family = binomial,
                  na.action = na.exclude,
                  data = df_reg_children)
```

```r
beta_ci_children <- confint(reg_children, parm="beta_",method="Wald")
beta_v_ci_children <- cbind(est=fixef(reg_children),beta_ci_children)
```

```r
or_v_ci_children <- exp(beta_v_ci_children)
```

```r
df_or_v_ci_children <- data.frame(or_v_ci_children)%>%
                              rownames_to_column(var = "Variable")%>%
                              mutate_at(vars(est:X97.5..),list(~round(.,3)))
df_or_v_ci_children
```

```r
colnames(df_or_v_ci_children) <- c("Variable","Estimate","Lower_bound","Upper_bound")
```

### Men

```r
reg_men <- glmer(Men_anemia ~ T_i30 +
                                P_tot_30 +
                                Current_age + N_year_educ + 
                                Smoker + Health_insurance + 
                                Religion_muslim + Religion_not_hindu_nor_muslim + 
                                Ethni_SC + Ethni_ST + Ethni_OBC +
                                Wealth_lowest + Wealth_second + Wealth_fourth + Wealth_highest + 
                                Urban +
                                Measured_month +
                                Measured_year +
                                (1|District_ID),
                  family = binomial,
                  na.action = na.exclude,
                  control = glmerControl(optimizer = "bobyqa"),                  
                  data = df_reg_men)
```

```r
beta_ci_men <- confint(reg_men, parm="beta_",method="Wald")
beta_v_ci_men <- cbind(est=fixef(reg_men),beta_ci_men)
```

```r
or_v_ci_men <- exp(beta_v_ci_men)
```

```r
df_or_v_ci_men <- data.frame(or_v_ci_men)%>%
                              rownames_to_column(var = "Variable")%>%
                              mutate_at(vars(est:X97.5..),list(~round(.,3)))
df_or_v_ci_men
```

## Plots

```r
df_plots <- rbind(df_or_v_ci_children%>%filter(Variable=="T_i30")%>%mutate(Group="Children"),
                  df_or_v_ci_men%>%filter(Variable=="T_i30")%>%mutate(Group="Men"),
                  df_or_v_ci_women%>%filter(Variable=="T_i30")%>%mutate(Group="Women")
                 )
df_plots
```

```r
options(repr.plot.width=8, repr.plot.height=4, scipen=10000)

plot_odds <- ggplot(df_plots,aes(y = Group, x = Estimate)) +
              geom_point(size = 3) +
              geom_errorbar(aes(xmin = Lower_bound, 
                                xmax = Upper_bound), 
                                height = 0.2) +
              coord_cartesian(x=c(0.9,1.2))+
              geom_vline(xintercept = 1, linetype = "dashed", color = "red") +  # Add a vertical line at x = 1
              labs(x = "Odds Ratio", y = "Group") +  # Label the axes
              scale_color_manual(values = c('red','blue'),
                                 na.value="white")+              
              labs(#title = "Odds ratios (95% confidence intervals) \n",
                   x = "\n Odds ratio (95% confidence interval)",
                   y = ""
                   ) +
              theme_minimal()+
              theme(plot.title = element_text(size = 20, face="bold"),
                    axis.text.x = element_text(size = 18,angle=45,hjust=1),
                    axis.text.y = element_text(size = 18),
                    axis.title.x = element_text(size = 20, hjust = 0.5),
                    axis.title.y = element_text(size = 20, hjust = 0.5)
                        )
plot_odds
```
