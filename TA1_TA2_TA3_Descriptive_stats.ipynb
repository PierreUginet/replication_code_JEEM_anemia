# Children

```r
df_children <- fread("./2_Regression_datasets/output/df_reg_children.csv")
sprintf("%i rows x %i columns", nrow(df_children), ncol(df_children))
head(df_children,2)
```

```r
df_reg_children <- df_children %>%
                mutate(T_i30_bin=factor(gsub("\\(","",T_i30_bin),levels=T_bins_labels))
```

```r
reg_children_T <- feols(Child_hemo_level_alti ~ i(T_i30_bin,ref=bin_ref_T) + P_tot_30 + 
                                        Child_female + Child_birth_order + Child_alive_age_month +
                                        Mother_prim_educ + Mother_second_educ + Mother_higher_educ +
                                        Mother_muslim + Mother_not_hindu_nor_muslim  + 
                                        Mother_ethni_SC + Mother_ethni_ST + Mother_ethni_OBC +
                                        Wealth_lowest + Wealth_second + Wealth_fourth + Wealth_highest + 
                                        Urban 
                                        |District_ID^Measured_month + District_ID^Measured_year,
                           vcov = ~District_ID,
                           df_reg_children)
```

```r
main_sample_children <- df_reg_children[obs(reg_children_T), ]
```

```r
main_sample_children <- data.frame(main_sample_children)
main_sample_children %>% 
    mutate(#Hemoglobin
           Child_anemia =  Children_anemia/1000,
           Child_severe_anemia = Child_severe_anemia/1000,
           Child_moderate_anemia = Child_moderate_anemia/1000,
           Child_mild_anemia = Child_mild_anemia/1000)%>%
    select(#Hemoglobin
           Child_hemo_level_alti,Child_anemia,Child_severe_anemia,Child_moderate_anemia,Child_mild_anemia,
           #Acute malnutrition
           Child_weight_for_height_zscore,
           #Child controls
           Child_female, Child_birth_order, Child_alive_age_month,
           #Diet
           Child_still_breastfeeding,
           Child_given_plain_water,Child_given_juice, Child_given_baby_formula,
           Child_given_fortified_baby_food, Child_given_soup, 
           Child_given_other_liquid, 
           #Animal proteins
           Child_given_milk,Child_given_yogurt,Child_given_cheese_yogurt_milk_products,
           Child_given_eggs,
           Child_given_chicken_duck_birds,Child_given_other_meat,
           Child_given_liver_heart_organs,Child_given_fish_sellfish,
           #Vegetables
           Child_given_bread_noodles_grains,Child_given_potatoes_cassava_tubers,
           Child_given_pumkins_carrots_squash, Child_given_green_vegetables,
           Child_given_mangoes_papaya_vitaminAfruits,Child_given_other_fruits,
           Child_given_beans_peas_lentils_nuts,
           Child_given_other_solid_semisolid_food,
           #Iron
           Child_iron_supplem_7d,
           #Bednet
           Bednet_slept,
           #Mother controls
           Mother_no_educ, Mother_prim_educ, Mother_second_educ, Mother_higher_educ,
           Mother_ethni_SC,Mother_ethni_ST,Mother_ethni_OBC,Mother_ethni_other,
           Mother_hindu, Mother_muslim, Mother_not_hindu_nor_muslim,
           #HH controls
           Wealth_lowest,Wealth_second,Wealth_middle,Wealth_fourth,Wealth_highest,
           HH_air_conditioner,
           Urban,
           #Water
           HH_time_water_source,
           Water_fetching_risk,
           #Malaria
           API_sup_1
          )%>%
    stargazer(nobs=TRUE, digits = 2,
              summary.stat = c("n","mean", "sd","min","p25","median","p75","max"),
              title = c("Summary statistics"),
              covariate.labels = c(#Hemoglobin
                                   "Hemoglobin level (g/L)",
                                   "Anemia (hemo <110)",
                                   "Severe anemia (hemo <70)",
                                   "Moderate anemia (70<hemo<100)",
                                   "Mild anemia (100<hemo<110)",
                                   #Acute malnutrition
                                   "Weight-for-height z-score (WHZ)",
                                    #Child controls
                                   "Female",
                                   "Birth order",
                                   "Age in months",
                                   #Diets
                                   "Still breastfed",
                                   "Diet (past 24h) : Plain water",
                                   "Diet (past 24h) : Juice", 
                                   "Diet (past 24h) : Baby formula",
                                   "Diet (past 24h) : Fortified baby food",
                                   "Diet (past 24h) : Soup",
                                   "Diet (past 24h) : Other liquid",
                                   "Diet (past 24h) : Milk (from animal)",
                                   "Diet (past 24h) : Yogurt",
                                   "Diet (past 24h) : Cheese, milk products",
                                   "Diet (past 24h) : Eggs",
                                   "Diet (past 24h) : Chicken, ducks birds",
                                   "Diet (past 24h) : Other meat",
                                   "Diet (past 24h) : Organs",
                                   "Diet (past 24h) : Fish, sellfish",
                                   "Diet (past 24h) : Bread, noodles, grains",
                                   "Diet (past 24h) : Potatoes, cassava, tubers",
                                   "Diet (past 24h) : Pumkins, carrots, squash",
                                   "Diet (past 24h) : Green vegetables",
                                   "Diet (past 24h) : Vitamin A fruits",
                                   "Diet (past 24h) : Other fruits",
                                   "Diet (past 24h) : Beans, peas, lentils, nuts",
                                   "Diet (past 24h) : Other semisolid food",
                                   #Iron
                                   "Iron supplementation in past 7 days",
                                   #Bednet
                                   "Slept under any bednet",
                                   #Mother controls
                                   "Mother no education",
                                   "Mother primary education",
                                   "Mother secondary education",
                                   "Mother higher education",
                                   "Mother scheduled caste (SC)",
                                   "Mother scheduled tribe (ST)",
                                   "Mother other backward caste (OBC)",
                                   "Mother not SC, ST or OBC",
                                   "Mother hindu",
                                   "Mother muslim",
                                   "Mother not hindu nor muslim",
                                   #HH controls
                                   "Wealth 1st quintile (poorest)",
                                   "Wealth 2nd quintile",
                                   "Wealth 3rd quintile",
                                   "Wealth 4th quintile",
                                   "Wealth 5th quintile (richest)",
                                   "Air conditioner",
                                   "Urban",
                                   #Water
                                   "Time to water source (minutes)",
                                   "Water fetching risk",
                                   #Malaria
                                   "API > 1"
                                   ),
              type="latex"
             )
```
