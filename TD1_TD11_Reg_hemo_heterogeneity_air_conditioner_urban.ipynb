```r
T_bins_labels <- c("<=17°C","17,21]","21,23]","23,25]","25,27]","27,29]","29,31]","31,33]",">33°C")
T_bins_labels_pretty <- c("≤17°C", "(17,21]", "(21,23]", "(23,25]", "(25,27]", "(27,29]", "(29,31]", "(31,33]", ">33°C")

bin_ref_T <-"17,21]"
```

```r
varT <- "T_i30_bin"

dict <- setNames(paste0("T: ",T_bins_labels_pretty), paste0(varT, "::", T_bins_labels))
```

## Air conditioner

```r
reg_children_AC <- feols(Child_hemo_level_alti ~ HH_air_conditioner*i(T_i30_bin,ref=bin_ref_T) + P_tot_30 + 
                                        Child_female + Child_birth_order + Child_alive_age_month + 
                                        Mother_prim_educ + Mother_second_educ + Mother_higher_educ +
                                        Mother_muslim + Mother_not_hindu_nor_muslim  + 
                                        Mother_ethni_SC + Mother_ethni_ST + Mother_ethni_OBC +
                                        Wealth_lowest + Wealth_second + Wealth_fourth + Wealth_highest + 
                                        Urban
                                        |District_ID^Measured_month + District_ID^Measured_year,
                        vcov = ~District_ID,
                    df_reg_children)
```

```r
reg_women_AC <- feols(Blood_hemo_level_alti ~ HH_air_conditioner*i(T_i30_bin,ref=bin_ref_T) + P_tot_30 + 
                                        Current_age + N_year_educ + Smoker + Health_insurance + 
                                        Religion_muslim + Religion_not_hindu_nor_muslim + 
                                        Ethni_SC + Ethni_ST + Ethni_OBC +
                                        Wealth_lowest + Wealth_second + Wealth_fourth + Wealth_highest + 
                                        Urban
                                        |District_ID^Measured_month + District_ID^Measured_year,
                        vcov = ~District_ID,
                    df_reg_women)
```

```r
reg_men_AC <- feols(Blood_hemo_level_alti ~ HH_air_conditioner*i(T_i30_bin,ref=bin_ref_T) + P_tot_30 + 
                                        Current_age + N_year_educ + Smoker + Health_insurance + 
                                        Religion_muslim + Religion_not_hindu_nor_muslim + 
                                        Ethni_SC + Ethni_ST + Ethni_OBC +
                                        Wealth_lowest + Wealth_second + Wealth_fourth + Wealth_highest + 
                                        Urban
                                        |District_ID^Measured_month + District_ID^Measured_year,
                        vcov = ~District_ID,
                    df_reg_men)
```

```r
etable(reg_children_AC,reg_women_AC,reg_men_AC,
       tex=TRUE,
       fitstat = ~ n + r2 + my,
       dict=c(dict,HH_air_conditioner= "Air conditioner"),
       drop="%!T_i30_bin|conditioner")
```
