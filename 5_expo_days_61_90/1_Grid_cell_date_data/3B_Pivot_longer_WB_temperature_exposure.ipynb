# Negative temperature

## Dataframe

```r
df_deg_bins_neg <- fread("./interm/df_WB_cells_CDC_negative.csv")
sprintf("%i x %i dataframe", nrow(df_deg_bins_neg), ncol(df_deg_bins_neg))
head(df_deg_bins_neg,2)
```

## Pivot longer dataframe

```r
df_deg_bins_piv_neg <- melt(df_deg_bins_neg%>%filter(!is.na(`T-1DayMin61`)), 
                       id.vars = c("Measured_date_CDC", "cell_x", "cell_y"),
                       variable.name = "Temp_bin",
                       value.name = "N_hours")
sprintf("%i x %i dataframe", nrow(df_deg_bins_piv_neg), ncol(df_deg_bins_piv_neg))
head(df_deg_bins_piv_neg,2)
```

```r
df_deg_bins_piv_neg_wo0 <- df_deg_bins_piv_neg[N_hours!=0]
```

# Positive temperature

## Dataframe

```r
df_deg_bins_pos <- fread("./interm/df_WB_cells_CDC_positive.csv")
sprintf("%i x %i dataframe", nrow(df_deg_bins_pos), ncol(df_deg_bins_pos))
head(df_deg_bins_pos,2)
```

## Pivot longer dataframe

```r
df_deg_bins_piv_pos <- melt(df_deg_bins_pos%>%filter(!is.na(T1DayMin61)), 
                       id.vars = c("Measured_date_CDC", "cell_x", "cell_y"),
                       variable.name = "Temp_bin",
                       value.name = "N_hours")
sprintf("%i x %i dataframe", nrow(df_deg_bins_piv_pos), ncol(df_deg_bins_piv_pos))
head(df_deg_bins_piv_pos,2)
```

```r
df_deg_bins_piv_pos_wo0 <- df_deg_bins_piv_pos[N_hours!=0]
```
