# Temperature bins

```r
temp_bins_N_days <- function(N_deg_bins,path_bins,pos){
    df_cells_date <- data.frame(df_DHS_distinct)
    for (i in if(pos){c(0:N_deg_bins)}else{c(0:-N_deg_bins-1)}){
        if(i < N_deg_bins & i != -N_deg_bins-1)
            {path_i <- paste0(path_bins,"t2m_bin_",i,"_",i+1,".csv")}
            else if(i==N_deg_bins & pos==TRUE)
                {path_i <- paste0(path_bins,"t2m_bin_",i,"_inf.csv")}
                else{path_i <-paste0(path_bins,"t2m_bin_-inf_",i+1,".csv")}
        df_Ti <- fread(path_i)%>%mutate(x=round(x,1),y=round(y,1))
        df_i <- df_cells_date %>% 
                    left_join(df_Ti,by=c("cell_x"="x","cell_y"="y"))
        names_i <- colnames(df_i)
        ncol_i <- ncol(df_i)
        # translator of measured date to column index
        date2index <- 1:ncol(df_i)	
        names(date2index) <- colnames(df_i)

        # Preallocate result matrix (rows = cells, cols = days)
        res_mat <- matrix(NA, nrow = nrow(df_i), ncol = 60)

        df_i_matrix <- as.matrix(df_i[, -c(1, 2, 3)])
        
        for (h in 1:30) {
          # Get the lagged date column names
          date_keys <- as.character(df_i$Measured_date_CDC - h - 60)
        
          # Convert to column indices relative to df_i_matrix
          idxs <- as.vector(date2index[date_keys]) - 3  # assumes first 3 cols were removed
        
          # Extract matrix values for the pairs (row_index, col_index) and create h-th column of res_mat
          res_mat[, h] <- df_i_matrix[cbind(1:nrow(df_i), idxs)]
        }
        gc()
        
        # Assign to df_i
        for (h in 1:30) {
          df_i[[ncol_i + h]] <- res_mat[, h]
        }
                                     
        #Change colnames
        newnames_i <- c()
        for (h in 1:30){
            newname <- paste0("T",i,"DayMin", 60 + h)
            newnames_i <- append(newnames_i,newname)
        }
        colnames(df_i) <- append(names_i,newnames_i)
        
        #Keep memory
        rm(df_Ti)
        gc()
        
        #Update dataframe
        df_cells_date <- df_i  %>%
                            select(-starts_with("4"))
        
        cat(paste0("Done: ",i, " out of ",N_deg_bins,"\n"))

    }
    return(df_cells_date)
}
```

## Positive air temperature

```r
path_T_pos <- "../.././0_input/one_degree_bins_T_WB/output/010_t2m_bins/t2m_bins/"
df_deg_bins_T_pos <- temp_bins_N_days(N_deg_bins=45,path_bins=path_T_pos,pos=TRUE)
sprintf("%i x %i dataframe", nrow(df_deg_bins_T_pos), ncol(df_deg_bins_T_pos))
head(df_deg_bins_T_pos,2)
```

```r
fwrite(df_deg_bins_T_pos,"./interm/df_T_cells_CDC_positive.csv")
```

## Negative air temperature

```r
path_T_neg <- "../.././0_input/one_degree_bins_T_WB/output/010_t2m_bins/t2m_bins/"
df_deg_bins_T_neg <- temp_bins_N_days(N_deg_bins=30,path_bins=path_T_neg,pos=FALSE)
sprintf("%i x %i dataframe", nrow(df_deg_bins_T_neg), ncol(df_deg_bins_T_neg))
head(df_deg_bins_T_neg,2)
```

```r
fwrite(df_deg_bins_T_neg,"./interm/df_T_cells_CDC_negative.csv")
```

```r
df_deg_bins_T <- df_deg_bins_T_neg %>%
                    left_join(df_deg_bins_T_pos, by=c("Measured_date_CDC","cell_x","cell_y"))
sprintf("%i x %i dataframe", nrow(df_deg_bins_T), ncol(df_deg_bins_T))
head(df_deg_bins_T,2)
```
