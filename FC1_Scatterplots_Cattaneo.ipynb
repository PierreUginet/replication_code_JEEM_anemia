```r
theme_perso <- function(...) {
                    theme_bw()+
                    theme(plot.title = element_text(size=30, face = "bold"),
                          axis.title.x = element_text(size = 20, face = "bold"),
                          axis.title.y = element_text(size = 20, face = "bold"),
                          axis.text.x =  element_text(size = 18),
                          axis.text.y =  element_text(size = 18))
    }
```

```r
scatterplot_children_T <- ggplot() + 
                                labs(x='X',y ='Y')+
                                # Add the dots
                                geom_point(data=result_children$data.dots, aes(x=x, y=fit), color="blue", size=2, shape='o')+
                                # Add the line
                                geom_line(data=result_children$data.line, aes(x=x, y=fit), color="blue", size=0.5)+
                                # Add the CI
                                geom_errorbar(data=result_children$data.ci, aes(x=x, ymin=ci.l, ymax=ci.r), color="blue", size=0.5, width = 0.02, linetype='solid')+
                                # Add the CB
                                geom_ribbon(data=result_children$data.cb, aes(x=x, ymin=cb.l, ymax=cb.r), fill="blue", alpha=0.2)+
                                # Add the polyreg
                                #geom_line(data=result$data.poly, aes(x=x, y=fit), color="red", size=0.5)+ 
                                labs(x="Air temperature (°C)",
                                     y="Hemoglobin level adjusted for altitude (g/L)\n ")+
                                theme_perso()
```

```r
scatterplot_children_WB <- ggplot() + 
                                labs(x='X',y ='Y')+
                                # Add the dots
                                geom_point(data=result_children_W$data.dots, aes(x=x, y=fit), color="blue", size=2, shape='o')+
                                # Add the line
                                geom_line(data=result_children_W$data.line, aes(x=x, y=fit), color="blue", size=0.5)+
                                # Add the CI
                                geom_errorbar(data=result_children_W$data.ci, aes(x=x, ymin=ci.l, ymax=ci.r), color="blue", size=0.5, width = 0.02, linetype='solid')+
                                # Add the CB
                                geom_ribbon(data=result_children_W$data.cb, aes(x=x, ymin=cb.l, ymax=cb.r), fill="blue", alpha=0.2)+
                                # Add the polyreg
                                #geom_line(data=result$data.poly, aes(x=x, y=fit), color="red", size=0.5)+ 
                                labs(x="Wet-bulb globe temperature (°C)",
                                     y="")+
                                theme_perso()
```

```r
options(repr.plot.width=15, repr.plot.height=7, scipen=9999)

plot_children <- annotate_figure(ggarrange(scatterplot_children_T,scatterplot_children_WB,nrow=1,ncol=2),
                                 top=text_grob("Children \n", face="bold", size=30))
plot_children
```
