# Merge DHS

```r
df_DHS <- bind_rows(df_DHS_15,df_DHS_19)
sprintf("%i x %i dataframe", nrow(df_DHS), ncol(df_DHS))
head(df_DHS,2)
```

```r
df_DHS_sel <- df_DHS %>%
                #Remove missing values
                filter(Usual_resident==1 & !is.na(State_2011) & !is.na(Measured_date_CDC) & 
                       !is.na(Child_hemo_level_alti))%>% 
                mutate(cell_x_01=round(cell_x_01,1),
                       cell_y_01=round(cell_y_01,1),
                       Wasted = case_when(is.na(Child_weight_for_height_zscore) ~ NA,
                                          Child_weight_for_height_zscore < -2 ~ 1000,
                                          TRUE ~ 0
                                          ),
                       Age_rounded = case_when(Child_alive_age_month<=6 ~ 6,
                                               Child_alive_age_month>6 & Child_alive_age_month<=12 ~ 12,
                                               Child_alive_age_month>12 & Child_alive_age_month<=18 ~ 18,
                                               Child_alive_age_month>18 & Child_alive_age_month<=24 ~ 24,
                                               Child_alive_age_month>24 & Child_alive_age_month<=30 ~ 30,
                                               Child_alive_age_month>30 & Child_alive_age_month<=36 ~ 36,
                                               Child_alive_age_month>36 & Child_alive_age_month<=42 ~ 42,
                                               Child_alive_age_month>42 & Child_alive_age_month<=48 ~ 48,
                                               Child_alive_age_month>48 & Child_alive_age_month<=54 ~ 54,
                                               Child_alive_age_month>54 & Child_alive_age_month<=59 ~ 59), 
                       Children_anemia = case_when(is.na(Child_hemo_level_alti) ~ NA,
                                                   Child_hemo_level_alti < 110 ~ 1000,
                                                   TRUE ~ 0
                                                        ),
                       Child_severe_anemia = case_when(is.na(Child_hemo_level_alti) ~ NA,
                                                        Child_hemo_level_alti < 70 ~ 1000,
                                                        Child_hemo_level_alti >= 70 ~ 0,
                                                        ),
                       Child_moderate_anemia = case_when(is.na(Child_hemo_level_alti) ~ NA,
                                                        Child_hemo_level_alti >= 70 & Child_hemo_level_alti <100 ~ 1000,
                                                        TRUE ~ 0
                                                        ),
                       Child_mild_anemia = case_when(is.na(Child_hemo_level_alti) ~ NA,
                                                        Child_hemo_level_alti >= 100 & Child_hemo_level_alti <110 ~ 1000,
                                                        TRUE ~ 0
                                                        ),
                       Child_no_anemia = case_when(is.na(Child_hemo_level_alti) ~ NA,
                                                        Child_hemo_level_alti >= 110 ~ 1000,
                                                        TRUE ~ 0
                                                        ), 
                       Season = case_when(Measured_month %in% c(12,1,2) ~ "Winter",
                                          Measured_month %in% 3:5 ~ "Summer",
                                          Measured_month %in% 6:8 ~ "Kharif",
                                          TRUE ~ "Rabi"),
                       Water_fetching_risk = case_when(is.na(HH_time_water_source) ~ NA,
                                                       HH_time_water_source>0 ~ 1,
                                                       TRUE ~ 0),
                       District_ID = paste0(State_2011,"_",District_2011))
sprintf("%i x %i dataframe", nrow(df_DHS_sel), ncol(df_DHS_sel))
head(df_DHS_sel,2)
```

# Malaria

```r
sf_malaria <- st_read(".././0_input/malaria/output/2011_Dist_malaria.gpkg")
```

```r
head(sf_malaria,2)
```

```r
sf_malaria_bin <- sf_malaria %>%
                    #st_drop_geometry()%>%
                    mutate(API_sup_1 = ifelse(API>1,1,ifelse(!is.na(API),0,NA)))
sprintf("%i x %i dataframe", nrow(sf_malaria_bin), ncol(sf_malaria_bin))
head(sf_malaria_bin,2)
```

```r
districts_telan <- unlist(df_DHS_15%>%
                                filter(State_2011=="Telangana")%>%distinct(District_2011),
                          use.names=FALSE)
districts_telan
```

```r
df_malaria <- sf_malaria_bin %>% 
                        st_drop_geometry()%>%
                        mutate(ST_NM = ifelse(ST_NM=="Andhra Pradesh" & DISTRICT %in% districts_telan, "Telangana",ST_NM))
sprintf("%i x %i dataframe", nrow(df_malaria), ncol(df_malaria))
head(df_malaria,2)
```

## Join to children measures

```r
df_DHS_final <- df_DHS_sel %>%
                        left_join(df_T_cells,by=c("Measured_date_CDC","cell_x_01"="cell_x","cell_y_01"="cell_y"))%>%
                        left_join(df_T_cells_hours,by=c("Measured_date_CDC","cell_x_01"="cell_x","cell_y_01"="cell_y"))%>%
                        left_join(df_T_cells_days,by=c("Measured_date_CDC","cell_x_01"="cell_x","cell_y_01"="cell_y"))%>%
                        left_join(df_WB_cells,by=c("Measured_date_CDC","cell_x_025"="cell_x","cell_y_025"="cell_y"))%>%
                        left_join(df_WB_cells_hours,by=c("Measured_date_CDC","cell_x_025"="cell_x","cell_y_025"="cell_y"))%>%
                        left_join(df_WB_cells_days,by=c("Measured_date_CDC","cell_x_025"="cell_x","cell_y_025"="cell_y"))%>%
                        left_join(df_P_cells,by=c("Measured_date_CDC","cell_x_025"="cell_x","cell_y_025"="cell_y"))%>%
                        left_join(df_malaria%>%select(ST_NM,DISTRICT,API_sup_1),
                                  by=c("State_2011"="ST_NM","District_2011"="DISTRICT"))%>%
                        filter(!is.na(Children_anemia))
sprintf("%i x %i dataframe", nrow(df_DHS_final), ncol(df_DHS_final))
head(df_DHS_final,2)
```
