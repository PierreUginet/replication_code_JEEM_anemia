# Merge DHS

```r
df_DHS <- bind_rows(df_DHS_15,df_DHS_19)
sprintf("%i x %i dataframe", nrow(df_DHS), ncol(df_DHS))
head(df_DHS,2)
```

```r
df_DHS_sel <- df_DHS %>%
                #Remove missing values
                filter(Usual_resident==1 & !is.na(State_2011) & !is.na(Measured_date_CDC))%>% 
                mutate(cell_x_01=round(cell_x_01,1),
                       cell_y_01=round(cell_y_01,1),
                       Measured_month = Interview_month,
                       Measured_year = Interview_year, 
                       Women_anemia = case_when(is.na(Blood_hemo_level_alti) ~ NA,
                                                Blood_hemo_level_alti <110 & Currently_pregnant==1 ~ 1000,
                                                Blood_hemo_level_alti<120 & Currently_pregnant==0 ~ 1000,
                                                TRUE ~ 0),
                       Women_severe_anemia = case_when(is.na(Blood_hemo_level_alti) ~ NA,
                                                       Blood_hemo_level_alti <70 & Currently_pregnant==1 ~ 1000,
                                                       Blood_hemo_level_alti<80 & Currently_pregnant==0 ~ 1000,
                                                       TRUE ~ 0),
                       Women_moderate_anemia = case_when(is.na(Blood_hemo_level_alti) ~ NA,
                                                         Blood_hemo_level_alti >= 70 & Blood_hemo_level_alti < 100 & Currently_pregnant==1 ~ 1000,
                                                         Blood_hemo_level_alti >= 80 & Blood_hemo_level_alti < 110 & Currently_pregnant==0 ~ 1000,
                                                         TRUE ~ 0),
                       Women_mild_anemia = case_when(is.na(Blood_hemo_level_alti) ~ NA,
                                                     Blood_hemo_level_alti >= 100 & Blood_hemo_level_alti < 110 & Currently_pregnant==1 ~ 1000,
                                                     Blood_hemo_level_alti >= 110 & Blood_hemo_level_alti < 120 & Currently_pregnant==0 ~ 1000,
                                                     TRUE ~ 0),
                       Women_no_anemia = case_when(is.na(Blood_hemo_level_alti) ~ NA,
                                                   Blood_hemo_level_alti >= 110 & Currently_pregnant==1 ~ 1000,
                                                   Blood_hemo_level_alti >= 120 & Currently_pregnant==0 ~ 1000,
                                                   TRUE ~ 0),
                       BMI_category = case_when(is.na(Body_mass_index) | Currently_pregnant==1 ~ NA,
                                                Body_mass_index <18.5 ~ "Underweight",
                                                Body_mass_index >=18.5 & Body_mass_index <25 ~ "Normal",
                                                Body_mass_index >= 25 & Body_mass_index <30 ~ "Overweight",
                                                Body_mass_index >=30 ~ "Obese"),
                       Blood_hypertension = case_when(is.na(Blood_pressure_systo) ~ NA,
                                                      Blood_pressure_systo >=140 |Blood_pressure_diasto >= 90 ~ 1000,
                                                      TRUE ~ 0),
                       Water_fetching_risk = case_when(is.na(HH_time_water_source) ~ NA,
                                                       HH_time_water_source>0 ~ 1,
                                                       TRUE ~ 0),
                       District_ID = paste0(State_2011,"_",District_2011))
sprintf("%i x %i dataframe", nrow(df_DHS_sel), ncol(df_DHS_sel))
head(df_DHS_sel,2)
```

# Malaria

```r
sf_malaria <- st_read("../0_input/malaria/output/2011_Dist_malaria.gpkg")
```

```r
df_malaria <- sf_malaria %>%
                    st_drop_geometry()%>%
                    mutate(API_sup_1 = ifelse(API>1,1,ifelse(!is.na(API),0,NA)),
                           ST_NM = ifelse(ST_NM=="Andhra Pradesh" & DISTRICT %in% districts_telan, "Telangana",ST_NM))
sprintf("%i x %i dataframe", nrow(df_malaria), ncol(df_malaria))
head(df_malaria,2)
```

## Join to women measures

```r
df_DHS_final <- df_DHS_sel %>%
                        left_join(df_T_cells,by=c("Measured_date_CDC","cell_x_01"="cell_x","cell_y_01"="cell_y"))%>%
                        left_join(df_T_cells_hours,by=c("Measured_date_CDC","cell_x_01"="cell_x","cell_y_01"="cell_y"))%>%
                        left_join(df_T_cells_days,by=c("Measured_date_CDC","cell_x_01"="cell_x","cell_y_01"="cell_y"))%>%
                        left_join(df_WB_cells,by=c("Measured_date_CDC","cell_x_025"="cell_x","cell_y_025"="cell_y"))%>%
                        left_join(df_WB_cells_hours,by=c("Measured_date_CDC","cell_x_025"="cell_x","cell_y_025"="cell_y"))%>%
                        left_join(df_WB_cells_days,by=c("Measured_date_CDC","cell_x_025"="cell_x","cell_y_025"="cell_y"))%>%
                        left_join(df_P_cells,by=c("Measured_date_CDC","cell_x_025"="cell_x","cell_y_025"="cell_y"))%>%
                        left_join(df_malaria%>%select(ST_NM,DISTRICT,API_sup_1),
                                  by=c("State_2011"="ST_NM","District_2011"="DISTRICT"))%>%
                        filter(!is.na(Women_anemia)&!is.na(T_i30)
                              )
sprintf("%i x %i dataframe", nrow(df_DHS_final), ncol(df_DHS_final))
head(df_DHS_final,2)
```
