# Heat index bar plot

```r
df_bar_plot_HI <- df_reg %>%
                        group_by(Measured_month)%>%
                        mutate(N_month=n())%>%
                        ungroup()%>%
                        filter(!is.na(WB_i30_bin))%>%
                        group_by(Measured_month,WB_i30_bin)%>%
                        summarize(value=n(),
                                  N_month=first(N_month))%>%
                        ungroup()%>%
                        mutate(WB_i30_bin=factor(WB_i30_bin,levels=c("<=14°C","(14,16]","(16,18]","(18,20]","(20,22]","(22,24]","(24,26]","(26,28]",">28°C")),
                               perc=value/N_month,
                               Perc=round(perc*100)
                               )
head(df_bar_plot_HI)
```

```r
options(repr.plot.width=10, repr.plot.height=10, scipen=9999)

plot_HI_month <- ggplot(df_bar_plot_HI,aes(x=as.factor(Measured_month),y=value,fill=WB_i30_bin))+
                            geom_bar(stat="identity") +    
                            theme_bw() +
                            labs(x="\n Measure Month",y="N children \n")+
                            scale_fill_manual(values=viridis(9),
                                              #labels=c("<50%","50-80","(75,90(","(90,95(",">95"),
                                              name="Wet-bulb T decile")+
                            geom_text(aes(x = as.factor(Measured_month), label = paste0(Perc,'%')),
                                      colour = 'black', position=position_stack(vjust=0.5))+
                            theme(legend.position="top",
                                  legend.title = element_text(size = 18,face ="bold"),
                                  legend.text = element_text(size = 16),
                                  axis.text.x = element_text(size = 16,face ="bold",angle=90),
                                  axis.text.y = element_text(size = 16),
                                  axis.title.x = element_text(size = 16, hjust = 0.5,face ="bold"),
                                  axis.title.y = element_text(size = 16, hjust = 0.5,face ="bold")
                                              )
plot_HI_month 
```

# Temperature barplot

```r
df_bar_plot_T <- df_reg %>%
                        group_by(Measured_month)%>%
                        mutate(N_month=n())%>%
                        ungroup()%>%
                        filter(!is.na(T_i30_bin))%>%
                        group_by(Measured_month,T_i30_bin)%>%
                        summarize(value=n(),
                                  N_month=first(N_month))%>%
                        ungroup()%>%
                        mutate(T_i30_bin=factor(T_i30_bin,levels=c("<=17°C","(17,21]",
                                                      "(21,23]","(23,25]","(25,27]","(27,29]","(29,31]","(31,33]",">33°C")),
                               perc=value/N_month,
                               Perc=round(perc*100)
                               )
head(df_bar_plot_T)
```

```r
options(repr.plot.width=10, repr.plot.height=10, scipen=9999)

plot_T_month <- ggplot(df_bar_plot_T,aes(x=as.factor(Measured_month),y=value,fill=T_i30_bin))+
                            geom_bar(stat="identity") +    
                            theme_bw() +
                            labs(x="\n Measure Month",y="N children \n")+
                            scale_fill_manual(values=viridis(9),
                                               #labels=c("<50%","50-75","(75,90(","(90,95(",">95"),
                                              name="Air T decile")+
                            geom_text(aes(x = as.factor(Measured_month), label = paste0(Perc,'%')),
                                      colour = 'black', position=position_stack(vjust=0.5))+
                            theme(legend.position="top",
                                  legend.title = element_text(size = 18,face ="bold"),
                                  legend.text = element_text(size = 16),
                                  axis.text.x = element_text(size = 16,face ="bold",angle=90),
                                  axis.text.y = element_text(size = 16),
                                  axis.title.x = element_text(size = 16, hjust = 0.5,face ="bold"),
                                  axis.title.y = element_text(size = 16, hjust = 0.5,face ="bold")
                                              )
plot_T_month 
```

```r
options(repr.plot.width=20, repr.plot.height=10, scipen=9999)

plot_both <- plot_T_month + plot_spacer() + plot_HI_month + plot_layout(widths=c(10,1,10))
plot_both
```

# Bar plot precipitation

```r
deciles_precip <- quantile(df_reg$P_tot_30, probs = seq(0, 1, by = 0.1))
deciles_precip
```

```r
df_bar_plot_precip <- df_reg %>%
                                mutate(P_cat_M3 = cut(P_tot_30, breaks = deciles_precip, 
                                                  labels = c("D1", "D2", "D3", "D4", "D5", "D6", "D7", "D8", "D9","D10"), include.lowest = TRUE))%>%
                                group_by(Measured_month)%>%
                                mutate(N_month=n())%>%
                                ungroup()%>%
                                group_by(Measured_month,P_cat_M3)%>%
                                summarize(N=n(),
                                          N_month=first(N_month))%>%
                                ungroup()%>%
                                mutate(perc=N/N_month,
                                       Perc=round(perc*100))
head(df_bar_plot_precip)
```

```r
options(repr.plot.width=10, repr.plot.height=10, scipen=9999)

plot_precip_month <- ggplot(df_bar_plot_precip,aes(x=as.factor(Measured_month),y=N,fill=P_cat_M3))+
                            geom_bar(stat="identity") +    
                            theme_bw() +
                            labs(x="\n Measure Month",y="Nb. of measured children \n")+
                            scale_fill_manual(values=viridis(10),
                            #                  labels=c('Yes',"No"),
                                              name="Precip deciles")+
                            geom_text(aes(x = as.factor(Measured_month), label = paste0(Perc,'%')),
                                      colour = 'black', position=position_stack(vjust=0.5))+
                            theme(legend.position="top",
                                  legend.title = element_text(size = 18,face ="bold"),
                                  legend.text = element_text(size = 16),
                                  axis.text.x = element_text(size = 16,face ="bold",angle=90),
                                  axis.text.y = element_text(size = 16),
                                  axis.title.x = element_text(size = 16, hjust = 0.5,face ="bold"),
                                  axis.title.y = element_text(size = 16, hjust = 0.5,face ="bold")
                                              )
plot_precip_month 
```

```r
options(repr.plot.width=25, repr.plot.height=10, scipen=9999)

plot_three <- plot_T_month + plot_spacer() + plot_HI_month + plot_spacer() + plot_precip_month + plot_layout(widths=c(10,1,10,1,10))
plot_three
```

```r
theme_perso <- function(...) {
                    theme_bw()+
                    theme(legend.position="none",
                          plot.title = element_text(size=30),
                          axis.title.x = element_text(size = 22, face ="bold"),
                          axis.text.x =  element_text(size = 20, angle=45, hjust=1),
                          axis.title.y = element_text(size = 22, face ="bold"),
                          axis.text.y =  element_text(size = 20)
                         )
}
```

```r
T_plot_mean <- ggplot(df_reg_boxplot_T_mean, aes(x=T_bin, y=value)) +
                geom_boxplot() +
                theme_perso() +
                ggtitle("Air temperature boxplot") +
                labs(y="Days in T bin (mean) \n")+
                scale_x_discrete(name="\n Air temperature bin",
                                 labels=c("<=17°C","(17,21]","(21,23]","(23,25]","(25,27]","(27,29]","(29,31]","(31,33]",">33°C")
                                )
```

```r
T_plot_max <- ggplot(df_reg_boxplot_T_max, aes(x=T_bin, y=value)) +
                geom_boxplot() +
                theme_perso() +
                ggtitle("Air temperature boxplot") +
                labs(y="Days in T bin (max) \n")+
                scale_x_discrete(name="\n Air temperature bin",
                                 labels=c("<=17°C","(17,21]","(21,23]","(23,25]","(25,27]","(27,29]","(29,31]","(31,33]",">33°C")
                                )
```
