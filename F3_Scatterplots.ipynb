# Scatterplots

```r
df_children_bins_T <- df_children_Tbin %>% 
                        arrange(T_i30)%>%
                        mutate(n20_T_i30=ntile(T_i30,100))%>%
                        group_by(n20_T_i30)%>%
                        summarize(T_i30_navg=mean(T_i30),
                                  Child_hemo_level_alti=mean(Child_hemo_level_alti,na.rm=TRUE))
```

```r
df_children_bins_WB <- df_children_Tbin %>% 
                        filter(!is.na(WB_i30))%>%
                        arrange(WB_i30)%>%
                        mutate(n20_WB_i30=ntile(WB_i30,100))%>%
                        group_by(n20_WB_i30)%>%
                        summarize(WB_i30_navg=mean(WB_i30),
                                  Child_hemo_level_alti=mean(Child_hemo_level_alti,na.rm=TRUE))
```

```r
df_women_bins_T <- df_women_Tbin %>% 
                        arrange(T_i30)%>%
                        mutate(n20_T_i30=ntile(T_i30,100))%>%
                        group_by(n20_T_i30)%>%
                        summarize(T_i30_navg=mean(T_i30),
                                  Blood_hemo_level_alti=mean(Blood_hemo_level_alti,na.rm=TRUE))
```

```r
df_women_bins_WB <- df_women_Tbin %>% 
                        filter(!is.na(WB_i30))%>%
                        arrange(WB_i30)%>%
                        mutate(n20_WB_i30=ntile(WB_i30,100))%>%
                        group_by(n20_WB_i30)%>%
                        summarize(WB_i30_navg=mean(WB_i30),
                                  Blood_hemo_level_alti=mean(Blood_hemo_level_alti,na.rm=TRUE))
```

```r
df_men_bins_T <- df_men_Tbin %>% 
                        arrange(T_i30)%>%
                        mutate(n20_T_i30=ntile(T_i30,100))%>%
                        group_by(n20_T_i30)%>%
                        summarize(T_i30_navg=mean(T_i30),
                                  Blood_hemo_level_alti=mean(Blood_hemo_level_alti,na.rm=TRUE))
```

```r
df_men_bins_WB <- df_men_Tbin %>% 
                        filter(!is.na(WB_i30))%>%
                        arrange(WB_i30)%>%
                        mutate(n20_WB_i30=ntile(WB_i30,100))%>%
                        group_by(n20_WB_i30)%>%
                        summarize(WB_i30_navg=mean(WB_i30),
                                  Blood_hemo_level_alti=mean(Blood_hemo_level_alti,na.rm=TRUE))
```

```r
theme_perso <- function(...) {
                    theme_bw()+
                    theme(plot.title = element_text(size=30, face = "bold"),
                          axis.title.x = element_text(size = 20, face = "bold"),
                          axis.title.y = element_text(size = 20, face = "bold"),
                          axis.text.x =  element_text(size = 18),
                          axis.text.y =  element_text(size = 18))
    }
```

```r
scatterplot_children_T <- ggplot(data = df_children_bins_T, aes(x = T_i30_navg, y = Child_hemo_level_alti)) +
                              geom_point()+
                              coord_cartesian(x=c(-3,37),y=c(95,112))+
                              labs(x="\n Air temperature (°C)",
                                   y="Hemoglobin level adjusted for altitude (g/L)\n ")+
                              theme_perso()
```

```r
scatterplot_children_WB <- ggplot(data = df_children_bins_WB, aes(x = WB_i30_navg, y = Child_hemo_level_alti)) +
                              geom_point()+
                              coord_cartesian(x=c(-3,37),y=c(95,112))+
                              labs(x="\n Wet-bulb globe temperature (°C)",
                                   y="")+
                              theme_perso()
```

```r
options(repr.plot.width=15, repr.plot.height=7, scipen=9999)

plot_children <- annotate_figure(ggarrange(scatterplot_children_T,scatterplot_children_WB,nrow=1,ncol=2),
                                 top=text_grob("Children \n", face="bold", size=30))
plot_children
```

```r
scatterplot_women_T <- ggplot(data = df_women_bins_T, aes(x = T_i30_navg, y = Blood_hemo_level_alti)) +
                          geom_point()+
                          coord_cartesian(x=c(-3,37),y=c(108,122))+
                          labs(x="\n Air temperature (°C)",
                               y="Hemoglobin level adjusted for altitude (g/L) \n")+
                            theme_perso()
```

```r
scatterplot_women_WB <- ggplot(data = df_women_bins_WB, aes(x = WB_i30_navg, y = Blood_hemo_level_alti)) +
                          geom_point()+
                          coord_cartesian(x=c(-3,37),y=c(108,122))+
                          labs(x="\n Wet-bulb globe temperature (°C)",
                               y="")+
                            theme_perso()
```

```r
options(repr.plot.width=15, repr.plot.height=7, scipen=9999)

plot_women <- annotate_figure(ggarrange(scatterplot_women_T,scatterplot_women_WB,nrow=1,ncol=2),
                                 top=text_grob("Women \n", face="bold", size=30))
plot_women
```

```r
scatterplot_men_T <- ggplot(data = df_men_bins_T, aes(x = T_i30_navg, y = Blood_hemo_level_alti)) +
                          geom_point()+
                          coord_cartesian(x=c(-3,37),y=c(132,145))+
                          labs(x="\n Air temperature (°C)",
                               y="Hemoglobin level adjusted for altitude (g/L) \n")+
                            theme_perso()
```

```r
scatterplot_men_WB <- ggplot(data = df_men_bins_WB, aes(x = WB_i30_navg, y = Blood_hemo_level_alti)) +
                          geom_point()+
                          coord_cartesian(x=c(-3,37),y=c(132,145))+
                          labs(x="\n Wet-bulb globe temperature (°C)",
                               y="")+
                            theme_perso()
```

```r
options(repr.plot.width=15, repr.plot.height=7, scipen=9999)

plot_men <- annotate_figure(ggarrange(scatterplot_men_T,scatterplot_men_WB,nrow=1,ncol=2),
                                 top=text_grob("Men \n", face="bold", size=30))
plot_men
```

```r
options(repr.plot.width=15, repr.plot.height=21, scipen=9999)

plot_all <- plot_children + plot_women + plot_men +  plot_layout(ncol=1)
plot_all
```
