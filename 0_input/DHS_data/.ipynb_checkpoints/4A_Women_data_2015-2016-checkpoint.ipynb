```r
path <- "../IA_2015-16_DHS/IAIR74DT_Women_Stata/IAIR74FL.DTA"
```

### Select columns of interest

```r
df_women <- haven::read_dta(path,
                            col_select=c(#Identifiers
                                         "caseid","v001","v002","v003","v006","v007","v016","v008",
                                         # Woman characteristics
                                         "v012","v013","v130","s116","v133","v463z","v481","v213",
                                         # Hemoglobin level
                                         "v456",
                                         # Other health outcomes
                                         "sb16s","sb16d","sb70","v445",
                                         # Nutrition
                                         "s726a","s726b","s726c","s726d","s726e","s726f","s726g","s726h","s726i",
                                         # Wealth
                                         "v190",
                                         # Residence
                                         "v025","v135"
                                       )
                            )
sprintf("%i x %i dataframe", nrow(df_women), ncol(df_women))
head(df_women,2)
```

# Creation of variables after reading labels

```r
df_women_var <- df_women %>%
                    mutate(DHS_round="2015", 
                           # Woman characteristics
                           Current_age = v012,
                           Age_group=haven::as_factor(v013),
                           Religion_hindu = ifelse(v130==1,1,0), 
                           Religion_muslim = ifelse(v130==2,1,0), 
                           Religion_not_hindu_nor_muslim = ifelse(v130!=1 & v130!=2,1,0),
                           Ethni_SC =ifelse(s116==1,1,
                                                  ifelse(s116==8|is.na(s116),NA,0)),
                           Ethni_ST =ifelse(s116==2,1,
                                                        ifelse(s116==8|is.na(s116),NA,0)),
                           Ethni_OBC =ifelse(s116==3,1,
                                                        ifelse(s116==8|is.na(s116),NA,0)),
                           Ethni_other =ifelse(s116==4,1,
                                                        ifelse(s116==8|is.na(s116),NA,0)),
                           N_year_educ = ifelse(v133==97,NA,v133), 
                           Smoker = ifelse(v463z==0,1,0),
                           Health_insurance  = v481,
                           Currently_pregnant = ifelse(v213==1,1,0),
                           # Hemoglobin level
                           Blood_hemo_level_alti = ifelse(v456==997,NA,v456),
                           # Other health outcomes
                           Blood_pressure_systo = ifelse(sb16s %in% c(994,995,996),NA, sb16s),
                           Blood_pressure_diasto = ifelse(sb16d %in% c(994,995,996),NA, sb16d),
                           Blood_glucose_level= ifelse(sb70 %in% c(995,996,998),NA,sb70),
                           # Body mass index (for non pregnant women)
                           Body_mass_index = ifelse(v445<1200 | v445>6000 | v213==1,NA,v445/100),
                           # Nutrition
                           Diet_milk_curd_daily = ifelse(s726a==1,1,0),
                           Diet_milk_curd_week = ifelse(s726a==2,1,0),
                           Diet_milk_curd_oc = ifelse(s726a==3,1,0),
                           Diet_milk_curd_never = ifelse(s726a==0,1,0),
                           Diet_pulses_beans_daily = ifelse(s726b==1,1,0),
                           Diet_pulses_beans_week = ifelse(s726b==2,1,0),
                           Diet_pulses_beans_oc = ifelse(s726b==3,1,0),
                           Diet_pulses_beans_never = ifelse(s726b==0,1,0),
                           Diet_green_veg_daily = ifelse(s726c==1,1,0),
                           Diet_green_veg_week = ifelse(s726c==2,1,0),
                           Diet_green_veg_oc = ifelse(s726c==3,1,0),
                           Diet_green_veg_never = ifelse(s726c==0,1,0),
                           Diet_fruits_daily = ifelse(s726d==1,1,0),
                           Diet_fruits_week = ifelse(s726d==2,1,0),
                           Diet_fruits_oc = ifelse(s726d==3,1,0),
                           Diet_fruits_never = ifelse(s726d==0,1,0),
                           Diet_eggs_daily = ifelse(s726e==1,1,0),
                           Diet_eggs_week = ifelse(s726e==2,1,0),
                           Diet_eggs_oc = ifelse(s726e==3,1,0),
                           Diet_eggs_never = ifelse(s726e==0,1,0),                           
                           Diet_fish_daily = ifelse(s726f==1,1,0),
                           Diet_fish_week = ifelse(s726f==2,1,0),
                           Diet_fish_oc = ifelse(s726f==3,1,0),
                           Diet_fish_never = ifelse(s726f==0,1,0), 
                           Diet_chicken_meat_daily = ifelse(s726g==1,1,0),
                           Diet_chicken_meat_week = ifelse(s726g==2,1,0),
                           Diet_chicken_meat_oc = ifelse(s726g==3,1,0),
                           Diet_chicken_meat_never = ifelse(s726g==0,1,0),  
                           Diet_fried_food_daily = ifelse(s726h==1,1,0),
                           Diet_fried_food_week = ifelse(s726h==2,1,0),
                           Diet_fried_food_oc = ifelse(s726h==3,1,0),
                           Diet_fried_food_never = ifelse(s726h==0,1,0), 
                           Diet_aerated_drinks_daily = ifelse(s726i==1,1,0),
                           Diet_aerated_drinks_week = ifelse(s726i==2,1,0),
                           Diet_aerated_drinks_oc = ifelse(s726i==3,1,0),
                           Diet_aerated_drinks_never = ifelse(s726i==0,1,0),
                           # Wealth
                           Wealth_lowest = ifelse(v190==1,1,0),
                           Wealth_second = ifelse(v190==2,1,0),
                           Wealth_middle = ifelse(v190==3,1,0),
                           Wealth_fourth = ifelse(v190==4,1,0),
                           Wealth_highest = ifelse(v190==5,1,0),
                           # Residence
                           Urban = ifelse(v025==1,1,0),
                           Usual_resident = ifelse(v135==2,0,v135)
                           )
```

## Select variables of interest

```r
df_women_select <- df_women_var %>%
                        select(# IDs
                               DHS_round,DHSCLUST=v001,HH_ID=v002,Women_id=caseid,Resp_nb=v003,
                               Interview_month=v006,Interview_year=v007,Interview_day=v016,
                               # Woman characteristics
                               Current_age, Age_group,
                               Religion_hindu, Religion_muslim, Religion_not_hindu_nor_muslim,
                               Ethni_SC,Ethni_ST,Ethni_OBC,Ethni_other,
                               N_year_educ,
                               Smoker,Health_insurance,
                               Currently_pregnant, 
                               # Hemoglobin level
                               Blood_hemo_level_alti,
                               # Other health outcomes
                               Blood_pressure_systo,Blood_pressure_diasto,Blood_glucose_level,
                               Body_mass_index,
                               # Nutrition
                               Diet_milk_curd_daily,Diet_milk_curd_week,Diet_milk_curd_oc,Diet_milk_curd_never,
                               Diet_pulses_beans_daily,Diet_pulses_beans_week,Diet_pulses_beans_oc,Diet_pulses_beans_never,
                               Diet_green_veg_daily,Diet_green_veg_week,Diet_green_veg_oc,Diet_green_veg_never,
                               Diet_fruits_daily,Diet_fruits_week,Diet_fruits_oc,Diet_fruits_never,
                               Diet_eggs_daily,Diet_eggs_week,Diet_eggs_oc,Diet_eggs_never,                           
                               Diet_fish_daily,Diet_fish_week,Diet_fish_oc,Diet_fish_never,
                               Diet_chicken_meat_daily,Diet_chicken_meat_week,Diet_chicken_meat_oc,Diet_chicken_meat_never,  
                               Diet_fried_food_daily,Diet_fried_food_week,Diet_fried_food_oc,Diet_fried_food_never, 
                               Diet_aerated_drinks_daily,Diet_aerated_drinks_week,Diet_aerated_drinks_oc,Diet_aerated_drinks_never,
                               # Wealth
                               Wealth_lowest,Wealth_second,Wealth_middle,Wealth_fourth,Wealth_highest,
                               # Residence
                               Urban,Usual_resident
                                )
sprintf("%i x %i dataframe", nrow(df_women_select), ncol(df_women_select))
head(df_women_select, 2)
```

## Add bednet & water access in HH members file

```r
path_HH <- "../IA_2015-16_DHS/IAPR74DT_HHmembers_Stata/IAPR74FL.DTA"
df_HH <- haven::read_dta(path_HH,
                         col_select=c("hvidx","ha0","hv001","hv002","hv003",
                                      # Bednet
                                      "hml12",
                                      # Water
                                      "hv204",
                                      # Air conditioner/cooler
                                      "sh37q",
                                      # WASH
                                      "hv237","hv205","hv201"
                                     )
                        )
sprintf("%i x %i dataframe", nrow(df_HH), ncol(df_HH))
head(df_HH, 2)
```

```r
df_watbed <- df_HH %>%
                mutate(Bednet_slept = ifelse(hml12==0,0,1),
                       HH_time_water_source = ifelse(hv204==996,0,
                                                    ifelse(hv204==998,NA,hv204)),
                       HH_air_conditioner = sh37q,
                       HH_treat_water = ifelse(hv237==8,NA,hv237),
                       HH_no_toilet = ifelse(hv205==30|hv205==31,1,0),
                       HH_well_water = ifelse(hv201==21|hv201==31|hv201==32,1,ifelse(!is.na(hv201),0,NA))
                       )%>%
                select(hvidx,hv001,hv002,
                       Bednet_slept,
                       HH_time_water_source,
                       HH_air_conditioner,
                       HH_treat_water,HH_no_toilet,HH_well_water)
sprintf("%i x %i dataframe", nrow(df_watbed), ncol(df_watbed))
head(df_watbed, 2)
```

```r
df_women_select_watbed <- df_women_select %>%
                            left_join(df_watbed,by=c("DHSCLUST"="hv001","HH_ID"="hv002","Resp_nb"="hvidx"))
sprintf("%i x %i dataframe", nrow(df_women_select_watbed), ncol(df_women_select_watbed))
head(df_women_select_watbed, 2)
```

## Add state and district according to 2011 census

```r
df_2011_census <- read_csv("./interm/Matching_DHSclusters_2015_census_2011.csv",show_col_types = FALSE)
sprintf("%i x %i dataframe", nrow(df_2011_census), ncol(df_2011_census))
head(df_2011_census, 2)
```

```r
df_women_census <- df_women_select_watbed %>%
                                left_join(df_2011_census,by="DHSCLUST")%>%
                                relocate(c(State_2011,District_2011),.after=DHSCLUST)
sprintf("%i x %i dataframe", nrow(df_women_census), ncol(df_women_census))
head(df_women_census, 2)
```

## Add cells ID

```r
df_cells_coord_01 <- read_csv("../ERA5/output/DHS_15_matching_clust_cells_0.1.csv",show_col_types = FALSE)
sprintf("%i x %i dataframe", nrow(df_cells_coord_01), ncol(df_cells_coord_01))
head(df_cells_coord_01, 2)
```

```r
df_cells_coord_025 <- read_csv("../ERA5/output/DHS_15_matching_clust_cells_0.25.csv",show_col_types = FALSE)
sprintf("%i x %i dataframe", nrow(df_cells_coord_025), ncol(df_cells_coord_025))
head(df_cells_coord_025, 2)
```

```r
df_women_cells <- df_women_census %>%
                                left_join(df_cells_coord_01%>%mutate(cell_x_01=cell_x,cell_y_01=cell_y),
                                          by="DHSCLUST")%>%
                                left_join(df_cells_coord_025%>%mutate(cell_x_025=cell_x,cell_y_025=cell_y),
                                          by="DHSCLUST")%>%
                                relocate(c(cell_x_01,cell_y_01,cell_x_025,cell_y_025),.after=District_2011)
sprintf("%i x %i dataframe", nrow(df_women_cells), ncol(df_women_cells))
head(df_women_cells, 2)
```

## Convert interview date in CDC

```r
df_women_CDC <- df_women_cells %>%
                        mutate(Measured_date = as.Date(paste(Interview_year, Interview_month, Interview_day, sep = "-")),
                               Measured_date_CDC = as.numeric(Measured_date - as.Date("1900-01-01")))%>%
                        relocate(c(Measured_date_CDC,Measured_date),.after=Interview_day)
sprintf("%i x %i dataframe", nrow(df_women_CDC), ncol(df_women_CDC))
head(df_women_CDC, 2)
```

## Final selection

```r
df_women_final <- df_women_CDC %>%
                            filter(!is.na(Measured_date_CDC) & !is.na(cell_x_01))
sprintf("%i x %i dataframe", nrow(df_women_final), ncol(df_women_final))
head(df_women_final, 2)
```
