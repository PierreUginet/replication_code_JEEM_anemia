## DHS clusters

```r
path_geo_15 <- "~/Dropbox/PSE_doctorant/Data/DHS/input/IA_2015-16_DHS/IAGE71FL_GeoData_shp/IAGE71FL.shp"
sf_DHS_15 <- st_read(path_geo_15)
```

```r
sf_DHS_15_valid <- sf_DHS_15%>%filter(SOURCE!="MIS")
nrow(sf_DHS_15_valid)
```

## Grid cells containing DHS clusters

## DHS 2015-16

```r
cells_DHS_15 <- terra::mask(cells, sf_DHS_15_valid)
```

```r
cells_DHS_15
```

#### Match DHS clusters to cells coordinates

```r
# Assign the point to the corresponding cell based on the point location within cells
cell_nums_15 <- cellFromXY(cells_DHS_15, sf_DHS_15_valid%>%st_drop_geometry()%>%select(LONGNUM,LATNUM))
```

```r
# Check which cells are outside mask (NA values)
cells_DHS_15_df <- as.data.frame(cells_DHS_15, xy = TRUE)%>%
                                    mutate(cell_id = rownames(.))
sprintf("%i x %i dataframe", nrow(cells_DHS_15_df), ncol(cells_DHS_15_df))
head(cells_DHS_15_df)
```

```r
# Assign cell indices to DHS clusters
sf_DHS_15_valid$cell_num <- cell_nums_15
```

```r
# Get the centroid coordinates of matched cells
cell_coords_15 <- xyFromCell(cells_DHS_15, cell_nums_15)
sf_DHS_15_valid$cell_x <- cell_coords_15[, 1]
sf_DHS_15_valid$cell_y <- cell_coords_15[, 2]
```

```r
DHS_within_cells_15 <- sf_DHS_15_valid%>%
                            st_drop_geometry()%>%
                            filter(cell_num %in% cells_DHS_15_df$cell_id)
sprintf("%i x %i dataframe", nrow(DHS_within_cells_15), ncol(DHS_within_cells_15))
head(DHS_within_cells_15)
```

```r
DHS_outside_cells_15 <- sf_DHS_15_valid%>%
                            st_drop_geometry()%>%
                            filter(!(cell_num %in% cells_DHS_15_df$cell_id))
sprintf("%i x %i dataframe", nrow(DHS_outside_cells_15), ncol(DHS_outside_cells_15))
head(DHS_outside_cells_15)
```

```r
# 1. Convert non-NA cells to SpatVector (centroids)
valid_cells_15 <- which(!is.na(values(cells_DHS_15)))
valid_points_15 <- xyFromCell(cells_DHS_15, valid_cells_15)
valid_vect_15 <- vect(valid_points_15, crs = crs(cells_DHS_15))
```

```r
# 2. Convert outside DHS points (sf) to SpatVector
outside_DHS_15_vect <- vect(DHS_outside_cells_15%>%st_as_sf(coords=c("LONGNUM","LATNUM"),crs=4326))
```

```r
# 3. Find nearest valid cell centroid for each DHS point
nearest_idx_15 <- terra::nearest(outside_DHS_15_vect, valid_vect_15)
```

```r
as.data.frame(nearest_idx_15)
```

```r
# 4. Assign those nearest cell indices
DHS_outside_cells_15_coord <- DHS_outside_cells_15%>%
                                    left_join(as.data.frame(nearest_idx_15),by=c("LONGNUM"="from_x","LATNUM"="from_y"))%>%
                                    select(-c(cell_num,cell_x,cell_y,from_id,distance))%>%
                                    dplyr::rename(cell_num=to_id,cell_x=to_x,cell_y=to_y)
sprintf("%i x %i dataframe", nrow(DHS_outside_cells_15_coord), ncol(DHS_outside_cells_15_coord))
head(DHS_outside_cells_15_coord)
```

### Save DHS 2015 with cells coordinates

```r
df_DHS_15_clust_cells_coord <- rbind(DHS_within_cells_15,DHS_outside_cells_15_coord)%>%
                                      select(DHSCLUST,cell_x,cell_y)#%>%
                                      #mutate(cell_x=round(cell_x,1),cell_y=round(cell_y,1))
sprintf("%i x %i dataframe", nrow(df_DHS_15_clust_cells_coord), ncol(df_DHS_15_clust_cells_coord))
head(df_DHS_15_clust_cells_coord)
```

```r
write_csv(df_DHS_15_clust_cells_coord,paste0("./output/DHS_15_matching_clust_cells_",reso,".csv"))
```

## Dataframe with cells containing all DHS clusters

```r
cells_all_DHS <- rbind(cells_DHS_15_df,cells_DHS_19_df)
sprintf("%i x %i dataframe", nrow(cells_all_DHS), ncol(cells_all_DHS))
```

```r
nrow(cells_all_DHS%>%distinct(x,y))
head(cells_all_DHS%>%distinct(x,y))
```

```r
df_clust_cells_coord <- rbind(df_DHS_15_clust_cells_coord,df_DHS_19_clust_cells_coord)
nrow(df_clust_cells_coord%>%distinct(cell_x,cell_y))
head(df_clust_cells_coord%>%distinct(cell_x,cell_y))
```

```r
missing_cells <- anti_join(df_clust_cells_coord, cells_all_DHS, by = c("cell_x"="x","cell_y"="y"))
sprintf("%i x %i dataframe", nrow(missing_cells), ncol(missing_cells))
head(missing_cells,2)
```

```r
ggplot() +
       geom_tile(data = missing_cells,
                      aes(x = cell_x, y = cell_y), fill = "red") + 
       coord_quickmap()+
       geom_sf(data = india,fill=NA, color="white", lwd=0.2, alpha=0.9)+
       theme_bw(base_family = "Roboto Condensed") +
       theme(legend.title = element_text(size = 16,face ="bold"),
             legend.text = element_text(size = 16),
             legend.position = 'top',
             axis.text.x = element_text(size = 16),
             axis.text.y = element_text(size = 16),
             axis.title.x = element_blank(),
             axis.title.y = element_blank())+
       labs(title="Missing cells")
```

```r
cells_all_DHS_no_dupli <- cells_all_DHS%>%
                                    group_by(x,y)%>%
                                    summarize(t2m_1=first(t2m_1), .groups = "drop")%>%
                                    ungroup()%>%
                                    select(-t2m_1)
sprintf("%i x %i dataframe", nrow(cells_all_DHS_no_dupli), ncol(cells_all_DHS_no_dupli))
head(cells_all_DHS_no_dupli,2)
```

```r
write_csv(cells_all_DHS_no_dupli, paste0("./output/Coord_cells_DHS_",reso,".csv"))
```
