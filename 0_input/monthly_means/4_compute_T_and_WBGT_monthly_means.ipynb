### For data at resolution 010

```python
# Collect all .nc file paths
nc_files = sorted(glob.glob(r"./ERA5_grid_cell_data/output/010_alltemps_*.nc"))
# Open and concatenate along time
all_temps_month = {}
for f in nc_files:
    ds = xr.open_mfdataset(f, combine='by_coords').load()
    daily = ds.resample(time="1ME").mean()
    all_temps_month[f]= daily
all_temps_month = xr.concat(all_temps_month.values(), dim='time')
print(all_temps_month)
```

```python
#turn the netcdf into a pandas dataframe and save it as a csv file

for var in ['t2m', 'tp']:
    df = all_temps_month[['time', 'lat', 'lon', var]].to_dataframe().pivot_table(index=['lat', 'lon'], columns=['time'], values=var)
    df.to_csv(f'./monthly_means/output/{var}_monthly_means.csv')
```
