# Women/men/children files

```r
directory <- "./output/men"
```

## List all .tiff files in the directory

```r
tiff_files <- list.files(directory, pattern = "\\.tif$", full.names = TRUE)
tiff_files
```

## Load the .tiff files as a raster stack

```r
raster_stack <- rast(tiff_files)
```

```r
raster_stack
```

## Sum the values across the stack

```r
raster_sum <- sum(raster_stack, na.rm = TRUE)
```

```r
raster_sum
```

## Save the output as a new .tiff file

```r
output_file <- paste0(directory,"/summed_men.tiff")
writeRaster(raster_sum, output_file, overwrite = TRUE)
```
