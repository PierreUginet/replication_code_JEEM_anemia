# Worldpop ([link](https://hub.worldpop.org/project/categories?id=8))

# India 100m Age structures in 2020

```r
children_files <- c("ind_f_5_2020_constrained","ind_m_5_2020_constrained")
```

```r
for (name in children_files){
    # Define the URL of the file to download
    file_url <- paste0("https://data.worldpop.org/GIS/AgeSex_structures/Global_2000_2020_Constrained/2020/IND//",name,".tif")
    # Define the folder and file name where you want to save it
    destination_folder <- "./output/children/"
    destination_file <- paste0(destination_folder,name,".tif")
    options(timeout = 300)
    curl_download(url = file_url, destfile = destination_file)
    raster_100m <- rast(destination_file)
    agg_factor <- 100 #target_resolution 10km
    raster_10km <- aggregate(raster_100m, fact = agg_factor, fun = sum, na.rm = TRUE)
    writeRaster(raster_10km, paste0(destination_folder,name,"_10km.tif"), overwrite = TRUE)
    file.remove(destination_file)
}
```
